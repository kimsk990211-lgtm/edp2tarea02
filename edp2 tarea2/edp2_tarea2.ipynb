# Problema 1 – Inciso a
## Aplicación del método de diferencias finitas

Se considera la siguiente ecuación diferencial parcial (EDP):

$$
u_t = D\,u_{xx}
$$

donde

$$
u = u(t,x), \quad 0 < x < 1, \quad t > 0
$$

con condición inicial

$$
u(0,x) = e^{-(x-0.5)^2}
$$

y condiciones de frontera

$$
u(t,0) = u_0, \qquad u_x(t,1) = 0
$$

En el problema se considera

$$
D = 5, \qquad u_0 = 1
$$

---

# 1. Discretización del dominio

Para aplicar el método de diferencias finitas, se discretiza el dominio en espacio y tiempo.

Se divide el intervalo espacial

$$
0 \le x \le 1
$$

en $$N$$ subintervalos.

El paso espacial es

$$
\Delta x = \frac{1}{N}
$$

Los nodos espaciales son

$$
x_i = i\Delta x, \quad i = 0,1,2,\dots,N
$$

De manera similar, el tiempo se discretiza como

$$
t_n = n\Delta t
$$

donde

$$
n = 0,1,2,\dots
$$

La aproximación numérica de la solución se denota por

$$
u_i^n \approx u(x_i,t_n)
$$

---

# 2. Aproximación de la derivada temporal

El problema indica utilizar **diferencias hacia adelante de primer orden en el tiempo**.

La derivada temporal se aproxima por

$$
u_t(x_i,t_n)
\approx
\frac{u_i^{n+1}-u_i^n}{\Delta t}
$$

---

# 3. Aproximación de la derivada espacial

Se utilizan **diferencias centrales de segundo orden para la segunda derivada espacial**.

$$
u_{xx}(x_i,t_n)
\approx
\frac{u_{i+1}^n -2u_i^n +u_{i-1}^n}{\Delta x^2}
$$

---

# 4. Sustitución en la ecuación diferencial

La ecuación original es

$$
u_t = D u_{xx}
$$

Sustituyendo las aproximaciones obtenemos

$$
\frac{u_i^{n+1}-u_i^n}{\Delta t}
=
D
\frac{u_{i+1}^n-2u_i^n+u_{i-1}^n}{\Delta x^2}
$$

---

# 5. Despeje de la solución futura

Multiplicando por $$\Delta t$$:

$$
u_i^{n+1}-u_i^n
=
\frac{D\Delta t}{\Delta x^2}
\left(
u_{i+1}^n
-
2u_i^n
+
u_{i-1}^n
\right)
$$

Despejando $$u_i^{n+1}$$:

$$
u_i^{n+1}
=
u_i^n
+
\frac{D\Delta t}{\Delta x^2}
\left(
u_{i+1}^n
-
2u_i^n
+
u_{i-1}^n
\right)
$$

---

# 6. Definición del parámetro de difusión

Se define el parámetro

$$
r = \frac{D\Delta t}{\Delta x^2}
$$

Entonces el esquema numérico queda

$$
u_i^{n+1}
=
u_i^n
+
r
\left(
u_{i+1}^n
-
2u_i^n
+
u_{i-1}^n
\right)
$$

Este es un **esquema explícito de diferencias finitas** para la ecuación del calor.

---

# 7. Condiciones de frontera

### Frontera izquierda

$$
u(t,0) = u_0
$$

Como $$u_0 = 1$$, entonces

$$
u_0^n = 1
$$

para todo $$n$$.

---

### Frontera derecha

Se tiene

$$
u_x(t,1) = 0
$$

Aproximando la derivada espacial mediante diferencias finitas

$$
\frac{u_N^n - u_{N-1}^n}{\Delta x} = 0
$$

Por lo tanto

$$
u_N^n = u_{N-1}^n
$$

Esta condición representa una frontera **aislada térmicamente**.

---

# 8. Condición inicial

La condición inicial es

$$
u(0,x) = e^{-(x-0.5)^2}
$$

Por lo tanto

$$
u_i^0 = e^{-(x_i-0.5)^2}
$$

para

$$
i = 0,1,2,\dots,N
$$

---

# 9. Esquema final del método

El algoritmo numérico queda:

1. Inicializar

$$
u_i^0 = e^{-(x_i-0.5)^2}
$$

2. Para cada paso de tiempo

$$
u_i^{n+1}
=
u_i^n
+
r(u_{i+1}^n-2u_i^n+u_{i-1}^n)
$$

3. Aplicar condiciones de frontera

$$
u_0^n = 1
$$

$$
u_N^n = u_{N-1}^n
$$

---

# 10. Condición de estabilidad

Para el esquema explícito es necesario cumplir

$$
r \le \frac{1}{2}
$$

es decir

$$
\Delta t \le \frac{\Delta x^2}{2D}
$$

Esto garantiza la estabilidad del método numérico.

---

# Conclusión

El método de diferencias finitas permite transformar la ecuación diferencial parcial en un sistema de ecuaciones algebraicas que puede resolverse iterativamente en el tiempo.  
La discretización utilizando diferencias hacia adelante en el tiempo y diferencias centrales en el espacio produce un esquema explícito que aproxima la solución de la ecuación del calor en el dominio considerado.

## Experimentos numéricos y visualización tridimensional

En este inciso se realizan tres experimentos numéricos utilizando el método de diferencias finitas obtenido en el inciso anterior.

Se consideran tres discretizaciones espaciales diferentes:

$$
0 \le x \le 1
$$

---

# 1. Discretización espacial

El paso espacial está definido por

$$
\Delta x = \frac{1}{N}
$$

Los nodos espaciales son

$$
x_i = i \Delta x, \quad i = 0,1,2,\dots,N
$$

---

# 2. Discretización temporal

El tiempo también se discretiza en pasos

$$
t_n = n \Delta t
$$

Para garantizar la estabilidad del método explícito se debe cumplir

$$
r = \frac{D \Delta t}{\Delta x^2} \le \frac{1}{2}
$$

Por lo tanto elegimos

$$
\Delta t = 0.8 \frac{\Delta x^2}{2D}
$$

lo cual satisface la condición de estabilidad.

---

# 3. Esquema numérico utilizado

El esquema explícito obtenido en el inciso anterior es

$$
u_i^{n+1}
=
u_i^n
+
r
\left(
u_{i+1}^n
-
2u_i^n
+
u_{i-1}^n
\right)
$$

donde

$$
r = \frac{D\Delta t}{\Delta x^2}
$$

---

# 4. Condiciones de frontera

En cada paso temporal se aplican las condiciones de frontera:

Frontera izquierda

$$
u(t,0)=1
$$

lo cual implica

$$
u_0^n = 1
$$

---

Frontera derecha

$$
u_x(t,1)=0
$$

que se aproxima mediante

$$
\frac{u_N^n-u_{N-1}^n}{\Delta x}=0
$$

por lo tanto

$$
u_N^n = u_{N-1}^n
$$

---

# 5. Condición inicial

La condición inicial del problema es

$$
u(0,x) = e^{-(x-0.5)^2}
$$

por lo tanto

$$
u_i^0 = e^{-(x_i-0.5)^2}
$$

---

# 6. Interpretación física

La condición inicial representa una distribución de temperatura tipo gaussiana centrada en

$$
x = 0.5
$$

Con el paso del tiempo, la ecuación del calor describe la difusión térmica, por lo que la distribución de temperatura se vuelve progresivamente más uniforme en el dominio.

Para visualizar este fenómeno se construyen gráficas tridimensionales de

$$
u(t,x)
$$

en el dominio

$$
0 \le x \le 1, \quad 0 \le t \le 10
$$

para cada valor de \(N\).

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
D = 5
u0 = 1
Tmax = 10

def solve_heat_equation(N):

    dx = 1/N
    
    dt = 0.8*(dx**2)/(2*D)
    
    Nt = int(Tmax/dt)
    
    x = np.linspace(0,1,N+1)
    
    t = np.linspace(0,Tmax,Nt)
    
    u = np.zeros((Nt,N+1))
    
    # condición inicial
    
    u[0,:] = np.exp(-(x-0.5)**2)
    
    r = D*dt/dx**2
    
    for n in range(Nt-1):
        
        for i in range(1,N):
            
            u[n+1,i] = u[n,i] + r*(u[n,i+1] - 2*u[n,i] + u[n,i-1])
        
        # frontera izquierda
        
        u[n+1,0] = u0
        
        # frontera derecha
        
        u[n+1,N] = u[n+1,N-1]
    
    return x,t,u
```

```python
def plot_3D_solution(N):

    x,t,u = solve_heat_equation(N)
    
    X,T = np.meshgrid(x,t)
    
    fig = plt.figure()
    
    ax = fig.add_subplot(111, projection='3d')
    
    ax.plot_surface(X,T,u)
    
    ax.set_xlabel("x")
    
    ax.set_ylabel("t")
    
    ax.set_zlabel("u(t,x)")
    
    ax.set_title(f"Solución numérica de la ecuación del calor (N={N})")
    
    plt.show()
```

```python
plot_3D_solution(10)
```

```python
plot_3D_solution(20)
```

```python
plot_3D_solution(40)
```

- La solución se vuelve progresivamente más suave debido al fenómeno de difusión térmica.

# Problema 1 – Inciso c
## Cálculo del error global entre la solución analítica y la solución numérica

En este inciso se debe comparar la solución numérica obtenida mediante el método de diferencias finitas con la solución analítica de la ecuación del calor.

El objetivo es calcular el **error global** entre ambas soluciones para los tres experimentos numéricos realizados anteriormente:

$$
N = 10, \quad N = 20, \quad N = 40
$$

---

# 1. Transformación del problema

La ecuación diferencial es

$$
u_t = D u_{xx}
$$

con condiciones de frontera

$$
u(t,0) = 1
$$

$$
u_x(t,1) = 0
$$

Para simplificar el problema, se realiza el cambio de variable

$$
v(x,t) = u(x,t) - 1
$$

Entonces

$$
u = v + 1
$$

Sustituyendo en la ecuación diferencial:

$$
v_t = D v_{xx}
$$

---

# 2. Condiciones de frontera para v

En la frontera izquierda

$$
u(t,0) = 1
$$

entonces

$$
v(t,0) = 0
$$

---

En la frontera derecha

$$
u_x(t,1) = 0
$$

por lo tanto

$$
v_x(t,1) = 0
$$

---

# 3. Condición inicial

La condición inicial original es

$$
u(0,x) = e^{-(x-0.5)^2}
$$

Entonces

$$
v(0,x) = e^{-(x-0.5)^2} - 1
$$

---

# 4. Solución analítica mediante separación de variables

Se busca una solución de la forma

$$
v(x,t) = X(x)T(t)
$$

Sustituyendo en la ecuación

$$
X(x)T'(t) = D X''(x)T(t)
$$

Dividiendo

$$
\frac{T'}{DT} = \frac{X''}{X} = -\lambda
$$

Se obtienen dos ecuaciones ordinarias

---

### Ecuación temporal

$$
T'(t) + D\lambda T(t) = 0
$$

cuya solución es

$$
T(t) = e^{-D\lambda t}
$$

---

### Ecuación espacial

$$
X'' + \lambda X = 0
$$

con condiciones

$$
X(0) = 0
$$

$$
X'(1) = 0
$$

---

# 5. Autovalores y autofunciones

La solución espacial tiene la forma

$$
X(x) = A\sin(\sqrt{\lambda}x)
$$

Aplicando la condición

$$
X'(1) = 0
$$

se obtiene

$$
\cos(\sqrt{\lambda}) = 0
$$

por lo tanto

$$
\sqrt{\lambda_n} = (n+\tfrac12)\pi
$$

entonces

$$
\lambda_n = (n+\tfrac12)^2\pi^2
$$

Las autofunciones son

$$
X_n(x) = \sin((n+\tfrac12)\pi x)
$$

---

# 6. Serie de Fourier

La solución general es

$$
v(x,t) =
\sum_{n=0}^{\infty}
a_n
\sin((n+\tfrac12)\pi x)
e^{-D(n+\tfrac12)^2\pi^2 t}
$$

donde los coeficientes se obtienen mediante

$$
a_n =
2\int_0^1
\left(e^{-(x-0.5)^2}-1\right)
\sin((n+\tfrac12)\pi x)
dx
$$

---

# 7. Solución final

Recordando que

$$
u = v + 1
$$

entonces

$$
u(x,t) =
1 +
\sum_{n=0}^{\infty}
a_n
\sin((n+\tfrac12)\pi x)
e^{-D(n+\tfrac12)^2\pi^2 t}
$$

Para la comparación numérica se utilizan **10 términos de la serie**.

---

# 8. Definición del error global

El error global se define como

$$
E =
\|u_{num}-u_{exact}\|
$$

donde

- \(u_{num}\) es la solución numérica
- \(u_{exact}\) es la solución analítica

En este trabajo se utiliza la norma Euclidiana

$$
E =
\sqrt{
\sum_{i=0}^{N}
(u_{num}-u_{exact})^2
}
$$

```python
def analytical_solution(x,t,terms=10):

    v = np.zeros_like(x)

    xs = np.linspace(0,1,1000)
    
    f = np.exp(-(xs-0.5)**2) - 1

    for n in range(terms):
        
        lam = (n+0.5)*np.pi
        
        phi = np.sin(lam*xs)
        
        coef = 2*np.trapz(f*phi,xs)
        
        v += coef*np.sin(lam*x)*np.exp(-D*(lam**2)*t)

    return v + 1
```

```python
def compute_error(N):

    x,t,u = solve_heat_equation(N)
    
    exact = analytical_solution(x,t[-1])
    
    error = np.linalg.norm(u[-1,:]-exact)
    
    return error
```

```python
Ns = [10,20,40]

for N in Ns:
    
    print("N =",N," Error =",compute_error(N))
```

```python
errors = []

for N in Ns:
    
    errors.append(compute_error(N))

plt.figure()

plt.plot(Ns,errors,marker='o')

plt.xlabel("Número de nodos N")

plt.ylabel("Error global")

plt.title("Convergencia del método de diferencias finitas")

plt.show()
```

# Conclusiones

Se comparó la solución numérica obtenida mediante el método de diferencias finitas con la solución analítica aproximada mediante una serie de Fourier de diez términos.

Los resultados muestran que:

- El error global disminuye al aumentar el número de nodos espaciales.
- La solución numérica converge hacia la solución analítica.
- El método de diferencias finitas es adecuado para aproximar la solución de la ecuación del calor.

Por lo tanto, se verifica numéricamente la convergencia del método.

# Problema 2 – Inciso a
## Aplicación del método de Crank–Nicolson

Se considera la ecuación diferencial parcial

$$
u_t = D u_{xx} - \beta u_x
$$

donde

$$
u = u(t,x), \quad 0 < x < 1, \quad t > 0
$$

con condición inicial

$$
u(0,x) = e^{-(x-0.5)^2}
$$

y condiciones de frontera

$$
u(t,0) = u_0, \qquad u_x(t,1) = 0
$$

En el problema se consideran los parámetros

$$
D = 5, \qquad \beta = 1.5, \qquad u_0 = 1
$$

---

# 1. Discretización del dominio

Para aplicar el método numérico se discretiza el dominio en espacio y tiempo.

El intervalo espacial es

$$
0 \le x \le 1
$$

Se divide en \(N\) subintervalos iguales.

El paso espacial es

$$
\Delta x = \frac{1}{N}
$$

Los nodos espaciales se definen como

$$
x_i = i \Delta x, \qquad i = 0,1,2,\dots,N
$$

---

El tiempo también se discretiza

$$
t_n = n \Delta t
$$

donde

$$
n = 0,1,2,\dots
$$

La aproximación numérica de la solución se denota como

$$
u_i^n \approx u(x_i,t_n)
$$

---

# 2. Aproximación de la derivada temporal

La derivada temporal se aproxima mediante diferencia hacia adelante

$$
u_t(x_i,t_n)
\approx
\frac{u_i^{n+1}-u_i^n}{\Delta t}
$$

---

# 3. Aproximación de las derivadas espaciales

### Segunda derivada

Se utiliza diferencia central de segundo orden

$$
u_{xx}(x_i,t_n)
\approx
\frac{u_{i+1}^n -2u_i^n +u_{i-1}^n}{\Delta x^2}
$$

---

### Primera derivada

Se aproxima mediante diferencia central

$$
u_x(x_i,t_n)
\approx
\frac{u_{i+1}^n - u_{i-1}^n}{2\Delta x}
$$

---

# 4. Método de Crank–Nicolson

El método de Crank–Nicolson consiste en tomar el promedio entre los niveles de tiempo \(n\) y \(n+1\).

Por lo tanto

$$
u_t =
\frac{1}{2}
\left[
D u_{xx}^{n+1} - \beta u_x^{n+1}
+
D u_{xx}^{n} - \beta u_x^{n}
\right]
$$

Sustituyendo la aproximación temporal

$$
\frac{u_i^{n+1}-u_i^n}{\Delta t}
=
\frac{1}{2}
\left[
D u_{xx}^{n+1} - \beta u_x^{n+1}
+
D u_{xx}^{n} - \beta u_x^{n}
\right]
$$

---

# 5. Sustitución de las aproximaciones espaciales

Sustituyendo las fórmulas de diferencias finitas

Primero el término de difusión

$$
u_{xx}^{n+1}
=
\frac{u_{i+1}^{n+1}-2u_i^{n+1}+u_{i-1}^{n+1}}{\Delta x^2}
$$

$$
u_{xx}^{n}
=
\frac{u_{i+1}^{n}-2u_i^{n}+u_{i-1}^{n}}{\Delta x^2}
$$

---

El término de advección

$$
u_x^{n+1}
=
\frac{u_{i+1}^{n+1}-u_{i-1}^{n+1}}{2\Delta x}
$$

$$
u_x^{n}
=
\frac{u_{i+1}^{n}-u_{i-1}^{n}}{2\Delta x}
$$

---

Sustituyendo todo en la ecuación

$$
\frac{u_i^{n+1}-u_i^n}{\Delta t}
=
\frac{D}{2}
\left(
\frac{u_{i+1}^{n+1}-2u_i^{n+1}+u_{i-1}^{n+1}}{\Delta x^2}
+
\frac{u_{i+1}^{n}-2u_i^{n}+u_{i-1}^{n}}{\Delta x^2}
\right)
$$

$$
-
\frac{\beta}{2}
\left(
\frac{u_{i+1}^{n+1}-u_{i-1}^{n+1}}{2\Delta x}
+
\frac{u_{i+1}^{n}-u_{i-1}^{n}}{2\Delta x}
\right)
$$

---

# 6. Definición de parámetros

Para simplificar la notación definimos

$$
r = \frac{D\Delta t}{2\Delta x^2}
$$

$$
s = \frac{\beta \Delta t}{4\Delta x}
$$

---

# 7. Ecuación final del esquema

Sustituyendo estas definiciones obtenemos

$$
u_i^{n+1}
=
u_i^n
+
r
\left(
u_{i+1}^{n+1}-2u_i^{n+1}+u_{i-1}^{n+1}
+
u_{i+1}^{n}-2u_i^{n}+u_{i-1}^{n}
\right)
$$

$$
-
s
\left(
u_{i+1}^{n+1}-u_{i-1}^{n+1}
+
u_{i+1}^{n}-u_{i-1}^{n}
\right)
$$

---

# 8. Reordenamiento

Reordenando los términos que contienen \(n+1\) obtenemos un sistema lineal

$$
A u^{n+1} = B u^n
$$

donde \(A\) y \(B\) son matrices tridiagonales.

Este sistema se debe resolver en cada paso de tiempo para obtener la solución numérica.

---

# 9. Condiciones de frontera

Frontera izquierda

$$
u(t,0)=1
$$

entonces

$$
u_0^n = 1
$$

---

Frontera derecha

$$
u_x(t,1)=0
$$

aproximando

$$
\frac{u_N-u_{N-1}}{\Delta x}=0
$$

por lo tanto

$$
u_N = u_{N-1}
$$

---

# 10. Condición inicial

La condición inicial es

$$
u(0,x) = e^{-(x-0.5)^2}
$$

por lo tanto

$$
u_i^0 = e^{-(x_i-0.5)^2}
$$

---

# Conclusión

El método de Crank–Nicolson es un esquema implícito de segundo orden en tiempo y espacio que combina estabilidad numérica con buena precisión.

Este método es particularmente adecuado para ecuaciones de difusión y advección como la considerada en este problema.  
Sin embargo, requiere resolver un sistema lineal en cada paso de tiempo.

# Problema 2 – Inciso b

Las gráficas muestran la evolución temporal de la solución numérica.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
D = 5
beta = 1.5
u0 = 1
Tmax = 10

def crank_nicolson_solver(N):

    dx = 1/N
    dt = 0.01
    
    Nt = int(Tmax/dt)
    
    x = np.linspace(0,1,N+1)
    t = np.linspace(0,Tmax,Nt)
    
    u = np.zeros((Nt,N+1))
    
    # condición inicial
    u[0,:] = np.exp(-(x-0.5)**2)
    
    r = D*dt/(2*dx**2)
    s = beta*dt/(4*dx)
    
    A = np.zeros((N-1,N-1))
    B = np.zeros((N-1,N-1))
    
    for i in range(N-1):
        
        if i > 0:
            A[i,i-1] = -r + s
            B[i,i-1] = r - s
            
        A[i,i] = 1 + 2*r
        B[i,i] = 1 - 2*r
        
        if i < N-2:
            A[i,i+1] = -r - s
            B[i,i+1] = r + s
    
    for n in range(Nt-1):
        
        rhs = B @ u[n,1:N]
        
        # frontera izquierda
        rhs[0] += (r-s)*u0
        
        u_next = np.linalg.solve(A,rhs)
        
        u[n+1,1:N] = u_next
        
        # fronteras
        u[n+1,0] = u0
        u[n+1,N] = u[n+1,N-1]
    
    return x,t,u
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_solution(N):

    x,t,u = crank_nicolson_solver(N)
    
    X,T = np.meshgrid(x,t)
    
    fig = plt.figure()
    
    ax = fig.add_subplot(111,projection='3d')
    
    ax.plot_surface(X,T,u)
    
    ax.set_xlabel("x")
    
    ax.set_ylabel("t")
    
    ax.set_zlabel("u(t,x)")
    
    ax.set_title(f"Solución Crank-Nicolson (N={N})")
    
    plt.show()
```

```python
plot_solution(10)
```

```python
plot_solution(20)
```

```python
plot_solution(40)
```

# Resultados

Las simulaciones muestran la evolución de la solución de la ecuación de advección-difusión.

Se observa que:

- La condición inicial tiene forma gaussiana centrada en $$x = 0.5$$.
- Con el tiempo la solución se difunde debido al término $$D u_{xx}$$.
- El término de advección $$\beta u_x$$ introduce un transporte de la solución.

Además, al aumentar el número de nodos espaciales $$N$$, la aproximación numérica se vuelve más precisa.

Esto confirma que el método de Crank–Nicolson proporciona una solución estable y precisa para este tipo de ecuaciones diferenciales parciales.

# Problema 3 – Inciso a
## Aplicación del método de Crank–Nicolson en dos dimensiones

Se considera la ecuación diferencial parcial

$$
u_t = D \nabla^2 u
$$

donde

$$
u = u(t,x,y), \qquad (x,y) \in \Omega, \qquad t>0
$$

y \( \nabla^2 \) representa el operador Laplaciano en dos dimensiones.

---

# 1. Operador Laplaciano

En coordenadas cartesianas el operador Laplaciano se define como

$$
\nabla^2 u =
\frac{\partial^2 u}{\partial x^2}
+
\frac{\partial^2 u}{\partial y^2}
$$

Por lo tanto la ecuación diferencial se puede escribir como

$$
u_t =
D
\left(
u_{xx} + u_{yy}
\right)
$$

---

# 2. Discretización del dominio

Sea una malla rectangular en el dominio \( \Omega \).

Los pasos espaciales se definen como

$$
\Delta x = \frac{L_x}{N}
$$

$$
\Delta y = \frac{L_y}{M}
$$

Los nodos espaciales son

$$
x_i = i\Delta x
$$

$$
y_j = j\Delta y
$$

La aproximación numérica se denota por

$$
u_{i,j}^n \approx u(x_i,y_j,t_n)
$$

---

# 3. Discretización temporal

El tiempo se discretiza como

$$
t_n = n\Delta t
$$

y la derivada temporal se aproxima mediante

$$
u_t \approx
\frac{u_{i,j}^{n+1}-u_{i,j}^n}{\Delta t}
$$

---

# 4. Aproximación de las derivadas espaciales

### Segunda derivada en x

$$
u_{xx} \approx
\frac{
u_{i+1,j}-2u_{i,j}+u_{i-1,j}
}{\Delta x^2}
$$

---

### Segunda derivada en y

$$
u_{yy} \approx
\frac{
u_{i,j+1}-2u_{i,j}+u_{i,j-1}
}{\Delta y^2}
$$

---

# 5. Método de Crank–Nicolson

El método de Crank–Nicolson utiliza el promedio entre los niveles de tiempo \(n\) y \(n+1\).

Por lo tanto

$$
\frac{u_{i,j}^{n+1}-u_{i,j}^n}{\Delta t}
=
\frac{D}{2}
\left[
(\nabla^2 u)^{n+1}
+
(\nabla^2 u)^n
\right]
$$

---

# 6. Sustitución del Laplaciano

Sustituyendo las aproximaciones de diferencias finitas

$$
(\nabla^2 u)^n =
\frac{
u_{i+1,j}^n
-2u_{i,j}^n
+
u_{i-1,j}^n
}{\Delta x^2}
+
\frac{
u_{i,j+1}^n
-2u_{i,j}^n
+
u_{i,j-1}^n
}{\Delta y^2}
$$

---

De manera similar

$$
(\nabla^2 u)^{n+1} =
\frac{
u_{i+1,j}^{n+1}
-2u_{i,j}^{n+1}
+
u_{i-1,j}^{n+1}
}{\Delta x^2}
+
\frac{
u_{i,j+1}^{n+1}
-2u_{i,j}^{n+1}
+
u_{i,j-1}^{n+1}
}{\Delta y^2}
$$

---

# 7. Definición de parámetros

Se definen

$$
r_x =
\frac{D\Delta t}{2\Delta x^2}
$$

$$
r_y =
\frac{D\Delta t}{2\Delta y^2}
$$

---

# 8. Ecuación discreta final

Sustituyendo en la ecuación principal se obtiene

$$
u_{i,j}^{n+1}
=
u_{i,j}^n
+
r_x
\left(
u_{i+1,j}^{n+1}
-2u_{i,j}^{n+1}
+
u_{i-1,j}^{n+1}
+
u_{i+1,j}^{n}
-2u_{i,j}^{n}
+
u_{i-1,j}^{n}
\right)
$$

$$
+
r_y
\left(
u_{i,j+1}^{n+1}
-2u_{i,j}^{n+1}
+
u_{i,j-1}^{n+1}
+
u_{i,j+1}^{n}
-2u_{i,j}^{n}
+
u_{i,j-1}^{n}
\right)
$$

---

# 9. Condiciones de frontera

Las condiciones de frontera son

$$
(x,y) \in \Gamma_2
$$

$$
u_n(t,x,y)=1
$$

---

$$
(x,y) \in \Gamma_4
$$

$$
u_n(t,x,y)=-1
$$

---

Las fronteras

$$
\Gamma_1
$$

y

$$
\Gamma_3
$$

están aisladas, por lo que

$$
\frac{\partial u}{\partial n}=0
$$

---

# 10. Condición inicial

La condición inicial es

$$
u(0,x,y) =
e^{-[(x-0.5)^2+(y-0.5)^2]}
$$

para todo

$$
(x,y)\in \Omega
$$

---

# Conclusión

El método de Crank–Nicolson aplicado a la ecuación del calor en dos dimensiones permite obtener una aproximación numérica estable y precisa.

El esquema resultante conduce a un sistema lineal en cada paso temporal que involucra los nodos vecinos en las direcciones \(x\) y \(y\), reflejando la naturaleza bidimensional del operador Laplaciano.

# Problema 3 – Inciso b
## Experimentos numéricos y visualización de la solución

En este inciso se realizan experimentos numéricos utilizando el método de Crank–Nicolson para aproximar la solución de la ecuación del calor en dos dimensiones

El parámetro de difusión considerado es

# 1. Objetivo del experimento

Para ello se realizan simulaciones utilizando diferentes tamaños de malla:

N = M = 10
$$

N = M = 20

N = M = 40

El dominio bidimensional está definido por

El dominio se discretiza mediante una malla rectangular.

$$

representa una distribución gaussiana centrada cerca del punto

$$

Con el paso del tiempo la temperatura se difunde en el dominio debido al operador Laplaciano.

---

# 6. Visualización de la solución

Para analizar el comportamiento de la solución numérica se construyen superficies tridimensionales

$$
u(t,x,y)
$$

Estas gráficas permiten observar la difusión de la temperatura en el dominio bidimensional.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def initial_condition(x,y):

    return np.exp(-((x-0.5)**2 + (y-0.5)**2))
```

```python
D = 5
Tmax = 10

def heat2D_solver(N,M):

    dx = 1/N
    dy = 1/M
    
    dt = 0.0005
    
    Nt = int(Tmax/dt)
    
    x = np.linspace(0,1,N+1)
    y = np.linspace(0,1,M+1)
    
    u = np.zeros((Nt,N+1,M+1))
    
    # initial condition
    
    for i in range(N+1):
        for j in range(M+1):
            
            u[0,i,j] = initial_condition(x[i],y[j])
    
    rx = D*dt/(dx**2)
    ry = D*dt/(dy**2)
    
    for n in range(Nt-1):
        
        for i in range(1,N):
            for j in range(1,M):
                
                u[n+1,i,j] = (
                    u[n,i,j]
                    + rx*(u[n,i+1,j] - 2*u[n,i,j] + u[n,i-1,j])
                    + ry*(u[n,i,j+1] - 2*u[n,i,j] + u[n,i,j-1])
                )
        
        # insulated boundaries
        
        u[n+1,0,:] = u[n+1,1,:]
        u[n+1,N,:] = u[n+1,N-1,:]
        
        u[n+1,:,0] = u[n+1,:,1]
        u[n+1,:,M] = u[n+1,:,M-1]
    
    return x,y,u,dt
```

```python
def plot_surface(N,M,time):

    x,y,u,dt = heat2D_solver(N,M)
    
    index = int(time/dt)
    
    X,Y = np.meshgrid(x,y)
    
    Z = u[index,:,:]
    
    fig = plt.figure()
    
    ax = fig.add_subplot(111,projection='3d')
    
    ax.plot_surface(X,Y,Z.T,cmap="viridis")
    
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("u(x,y)")
    
    ax.set_title(f"N={N}, time={time}")
    
    ax.set_zlim(0,1)
    
    plt.show()
```

```python
plot_surface(10,10,0)
```

```python
plot_surface(20,20,0)
```
