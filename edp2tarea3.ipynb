# Ejercicio 2.3

Usar las series de Taylor para obtener un conjunto de ecuaciones lineales para las constantes $A$, $B$, $C$, $D$ y $E$ en la expansión

$$
\frac{\partial^2 \phi}{\partial x^2}
=
A\phi_{i-2,j}+B\phi_{i-1,j}+C\phi_{i+1,j}+D\phi_{i+2,j}+E
$$

---

# Observación importante

Al revisar cuidadosamente la expresión del problema, se observa que el último término aparece como $+E$, es decir, como una constante independiente.

Sin embargo, para obtener una aproximación por diferencias finitas de $\phi_{xx}$ mediante expansiones de Taylor, la forma correcta normalmente debe ser

$$
\frac{\partial^2 \phi}{\partial x^2}
=
A\phi_{i-2,j}+B\phi_{i-1,j}+E\phi_{i,j}+C\phi_{i+1,j}+D\phi_{i+2,j}.
$$

Esto ocurre porque, al expandir los términos por Taylor, aparece un término proporcional a $\phi_{i,j}$ que no puede cancelarse con una constante pura $E$.

Por esta razón, en la presente solución se interpretará que el enunciado contiene una omisión tipográfica, y se trabajará con la forma estándar:

$$
\frac{\partial^2 \phi}{\partial x^2}
=
A\phi_{i-2,j}+B\phi_{i-1,j}+E\phi_{i,j}+C\phi_{i+1,j}+D\phi_{i+2,j}.
$$

---

# 1. Expansiones de Taylor alrededor del punto $(i,j)$

Sea una malla uniforme en la dirección $x$ con tamaño de paso $h$. Entonces:

$$
\phi_{i\pm1,j}=\phi(x_i\pm h,y_j), \qquad
\phi_{i\pm2,j}=\phi(x_i\pm 2h,y_j)
$$

Se desarrollan en serie de Taylor alrededor del punto $(x_i,y_j)$.

Para simplificar la notación, escribimos:

$$
\phi=\phi_{i,j}, \qquad
\phi_x=\left.\frac{\partial \phi}{\partial x}\right|_{(i,j)}, \qquad
\phi_{xx}=\left.\frac{\partial^2 \phi}{\partial x^2}\right|_{(i,j)},
$$

y de forma análoga para derivadas de orden superior.

## 1.1 Expansión de $\phi_{i-2,j}$

$$
\phi_{i-2,j}
=
\phi
-2h\phi_x
+\frac{(2h)^2}{2!}\phi_{xx}
-\frac{(2h)^3}{3!}\phi_{xxx}
+\frac{(2h)^4}{4!}\phi_{xxxx}
+O(h^5)
$$

Simplificando:

$$
\phi_{i-2,j}
=
\phi
-2h\phi_x
+2h^2\phi_{xx}
-\frac{4}{3}h^3\phi_{xxx}
+\frac{2}{3}h^4\phi_{xxxx}
+O(h^5)
$$

## 1.2 Expansión de $\phi_{i-1,j}$

$$
\phi_{i-1,j}
=
\phi
-h\phi_x
+\frac{h^2}{2}\phi_{xx}
-\frac{h^3}{6}\phi_{xxx}
+\frac{h^4}{24}\phi_{xxxx}
+O(h^5)
$$

## 1.3 Expansión de $\phi_{i+1,j}$

$$
\phi_{i+1,j}
=
\phi
+h\phi_x
+\frac{h^2}{2}\phi_{xx}
+\frac{h^3}{6}\phi_{xxx}
+\frac{h^4}{24}\phi_{xxxx}
+O(h^5)
$$

## 1.4 Expansión de $\phi_{i+2,j}$

$$
\phi_{i+2,j}
=
\phi
+2h\phi_x
+2h^2\phi_{xx}
+\frac{4}{3}h^3\phi_{xxx}
+\frac{2}{3}h^4\phi_{xxxx}
+O(h^5)
$$

---

# 2. Sustitución en la combinación lineal

Se desea que

$$
\phi_{xx}
=
A\phi_{i-2,j}+B\phi_{i-1,j}+E\phi_{i,j}+C\phi_{i+1,j}+D\phi_{i+2,j}
$$

Sustituyendo las expansiones de Taylor:

$$
\begin{aligned}
A\phi_{i-2,j}
&=
A\left(
\phi
-2h\phi_x
+2h^2\phi_{xx}
-\frac{4}{3}h^3\phi_{xxx}
+\frac{2}{3}h^4\phi_{xxxx}
\right), \\[1em]
B\phi_{i-1,j}
&=
B\left(
\phi
-h\phi_x
+\frac{h^2}{2}\phi_{xx}
-\frac{h^3}{6}\phi_{xxx}
+\frac{h^4}{24}\phi_{xxxx}
\right), \\[1em]
E\phi_{i,j}
&=
E\phi, \\[1em]
C\phi_{i+1,j}
&=
C\left(
\phi
+h\phi_x
+\frac{h^2}{2}\phi_{xx}
+\frac{h^3}{6}\phi_{xxx}
+\frac{h^4}{24}\phi_{xxxx}
\right), \\[1em]
D\phi_{i+2,j}
&=
D\left(
\phi
+2h\phi_x
+2h^2\phi_{xx}
+\frac{4}{3}h^3\phi_{xxx}
+\frac{2}{3}h^4\phi_{xxxx}
\right).
\end{aligned}
$$

Ahora se agrupan términos semejantes:

$$
\begin{aligned}
&A\phi_{i-2,j}+B\phi_{i-1,j}+E\phi_{i,j}+C\phi_{i+1,j}+D\phi_{i+2,j} \\
&=
(A+B+C+D+E)\phi \\[0.5em]
&\quad+
(-2A-B+C+2D)h\phi_x \\[0.5em]
&\quad+
\left(2A+\frac{B}{2}+\frac{C}{2}+2D\right)h^2\phi_{xx} \\[0.5em]
&\quad+
\left(
-\frac{4}{3}A-\frac{1}{6}B+\frac{1}{6}C+\frac{4}{3}D
\right)h^3\phi_{xxx} \\[0.5em]
&\quad+
\left(
\frac{2}{3}A+\frac{1}{24}B+\frac{1}{24}C+\frac{2}{3}D
\right)h^4\phi_{xxxx}
+O(h^5)
\end{aligned}
$$

---

# 3. Igualación de coeficientes

Para que la expresión anterior sea una aproximación de $\phi_{xx}$, deben cumplirse las siguientes condiciones:

- El coeficiente de $\phi$ debe ser cero.
- El coeficiente de $\phi_x$ debe ser cero.
- El coeficiente de $\phi_{xx}$ debe ser uno.
- El coeficiente de $\phi_{xxx}$ debe ser cero.
- El coeficiente de $\phi_{xxxx}$ debe ser cero.

Por lo tanto, se obtiene el siguiente sistema de ecuaciones lineales:

$$
A+B+C+D+E=0
$$

$$
-2A-B+C+2D=0
$$

$$
\left(2A+\frac{B}{2}+\frac{C}{2}+2D\right)h^2=1
$$

Equivalentemente,

$$
2A+\frac{B}{2}+\frac{C}{2}+2D=\frac{1}{h^2}
$$

Además,

$$
-\frac{4}{3}A-\frac{1}{6}B+\frac{1}{6}C+\frac{4}{3}D=0
$$

y

$$
\frac{2}{3}A+\frac{1}{24}B+\frac{1}{24}C+\frac{2}{3}D=0
$$

Este es el conjunto de ecuaciones lineales que pide el ejercicio.

---

# 4. Resolución del sistema

Por la simetría del stencil, se propone:

$$
A=D, \qquad B=C
$$

## 4.1 Verificación en la ecuación del término $\phi_x$

La segunda ecuación es

$$
-2A-B+C+2D=0
$$

Sustituyendo $A=D$ y $B=C$:

$$
-2A-B+B+2A=0
$$

lo cual se cumple automáticamente.

## 4.2 Verificación en la ecuación del término $\phi_{xxx}$

La cuarta ecuación es

$$
-\frac{4}{3}A-\frac{1}{6}B+\frac{1}{6}C+\frac{4}{3}D=0
$$

Sustituyendo $A=D$ y $B=C$:

$$
-\frac{4}{3}A-\frac{1}{6}B+\frac{1}{6}B+\frac{4}{3}A=0
$$

también se cumple automáticamente.

Entonces solo quedan por resolver las ecuaciones correspondientes a $\phi_{xx}$, $\phi_{xxxx}$ y $\phi$.

---

## 4.3 Ecuación del término $\phi_{xxxx}$

Partimos de

$$
\frac{2}{3}A+\frac{1}{24}B+\frac{1}{24}C+\frac{2}{3}D=0
$$

Usando $A=D$ y $B=C$:

$$
\frac{2}{3}A+\frac{1}{24}B+\frac{1}{24}B+\frac{2}{3}A=0
$$

$$
\frac{4}{3}A+\frac{1}{12}B=0
$$

Multiplicando por $12$:

$$
16A+B=0
$$

Por tanto,

$$
B=-16A
$$

---

## 4.4 Ecuación del término $\phi_{xx}$

Partimos de

$$
2A+\frac{B}{2}+\frac{C}{2}+2D=\frac{1}{h^2}
$$

Usando $A=D$ y $B=C$:

$$
2A+\frac{B}{2}+\frac{B}{2}+2A=\frac{1}{h^2}
$$

$$
4A+B=\frac{1}{h^2}
$$

Sustituyendo $B=-16A$:

$$
4A-16A=\frac{1}{h^2}
$$

$$
-12A=\frac{1}{h^2}
$$

Por lo tanto,

$$
A=-\frac{1}{12h^2}
$$

Como $A=D$, se obtiene

$$
D=-\frac{1}{12h^2}
$$

Y como $B=-16A$:

$$
B=-16\left(-\frac{1}{12h^2}\right)=\frac{16}{12h^2}=\frac{4}{3h^2}
$$

Como $B=C$, se obtiene

$$
C=\frac{4}{3h^2}
$$

---

## 4.5 Cálculo de $E$

De la primera ecuación:

$$
A+B+C+D+E=0
$$

Sustituyendo los valores hallados:

$$
-\frac{1}{12h^2}+\frac{4}{3h^2}+\frac{4}{3h^2}-\frac{1}{12h^2}+E=0
$$

Agrupando:

$$
\left(-\frac{1}{12}-\frac{1}{12}+\frac{4}{3}+\frac{4}{3}\right)\frac{1}{h^2}+E=0
$$

$$
\left(-\frac{2}{12}+\frac{8}{3}\right)\frac{1}{h^2}+E=0
$$

$$
\left(-\frac{1}{6}+\frac{16}{6}\right)\frac{1}{h^2}+E=0
$$

$$
\frac{15}{6}\frac{1}{h^2}+E=0
$$

$$
\frac{5}{2h^2}+E=0
$$

Por consiguiente,

$$
E=-\frac{5}{2h^2}
$$

---

# 5. Resultado final

Los coeficientes buscados son

$$
A=D=-\frac{1}{12h^2}
$$

$$
B=C=\frac{4}{3h^2}
$$

$$
E=-\frac{5}{2h^2}
$$

Por lo tanto, la aproximación final de diferencias finitas para la segunda derivada es

$$
\frac{\partial^2 \phi}{\partial x^2}
\approx
-\frac{1}{12h^2}\phi_{i-2,j}
+\frac{4}{3h^2}\phi_{i-1,j}
-\frac{5}{2h^2}\phi_{i,j}
+\frac{4}{3h^2}\phi_{i+1,j}
-\frac{1}{12h^2}\phi_{i+2,j}
$$

Sacando factor común $\frac{1}{12h^2}$, también puede escribirse como

$$
\frac{\partial^2 \phi}{\partial x^2}
\approx
\frac{-\phi_{i-2,j}+16\phi_{i-1,j}-30\phi_{i,j}+16\phi_{i+1,j}-\phi_{i+2,j}}{12h^2}
$$

---

# 6. Conclusión

Mediante el desarrollo en series de Taylor de los puntos vecinos $\phi_{i-2,j}$, $\phi_{i-1,j}$, $\phi_{i+1,j}$ y $\phi_{i+2,j}$ alrededor del punto central $(i,j)$, se obtuvo un sistema de ecuaciones lineales para los coeficientes $A$, $B$, $C$, $D$ y $E$.

Al resolver dicho sistema, se obtiene la fórmula centrada de cinco puntos para aproximar la derivada segunda respecto de $x$, la cual es una fórmula clásica de alta precisión en diferencias finitas.

---

# 7. Nota sobre el enunciado original

Si se usa literalmente la expresión

$$
\frac{\partial^2 \phi}{\partial x^2}
=
A\phi_{i-2,j}+B\phi_{i-1,j}+C\phi_{i+1,j}+D\phi_{i+2,j}+E
$$

con $E$ como constante independiente, entonces la expansión de Taylor produce un término en $\phi_{i,j}$ que no puede cancelarse correctamente con una constante pura.

Por ello, para que el método sea consistente, se interpreta que el término correcto es $E\phi_{i,j}$.

# Ejercicio 2.4

Construir una malla en el plano $(x,y)$ con incrementos $h$ en la dirección $x$ y $k$ en la dirección $y$.  
En el punto $(i,j)$ obtener expresiones en diferencias finitas para:

1. $\displaystyle \frac{\partial^2 \phi}{\partial x \partial y}$
2. $\displaystyle \frac{\partial^2 \phi}{\partial x^2} + x\frac{\partial^2 \phi}{\partial y^2}$

correctas hasta $O(\max(h^2,k^2))$.

---

# 1. Malla

Se define:

$$
x_i = x_0 + ih, \quad y_j = y_0 + jk
$$

$$
\phi_{i,j} = \phi(x_i,y_j)
$$

---

# 2. Aproximación de $\frac{\partial^2 \phi}{\partial x \partial y}$

Usamos los 4 puntos diagonales:

$$
\phi_{i+1,j+1}, \quad \phi_{i+1,j-1}, \quad \phi_{i-1,j+1}, \quad \phi_{i-1,j-1}
$$

---

## Taylor (idea clave)

Ejemplo:

$$
\phi_{i+1,j+1}
=
\phi + h\phi_x + k\phi_y + hk\phi_{xy} + \frac{h^2}{2}\phi_{xx} + \frac{k^2}{2}\phi_{yy} + \cdots
$$

(Lo mismo para los otros 3 puntos)

---

## Combinación

Consideramos:

$$
\phi_{i+1,j+1} - \phi_{i+1,j-1} - \phi_{i-1,j+1} + \phi_{i-1,j-1}
$$

👉 Todos los términos se cancelan excepto:

$$
4hk \phi_{xy}
$$

---

## Resultado

$$
\boxed{
\left(\frac{\partial^2 \phi}{\partial x \partial y}\right)_{i,j}
\approx
\frac{
\phi_{i+1,j+1}
-\phi_{i+1,j-1}
-\phi_{i-1,j+1}
+\phi_{i-1,j-1}
}{4hk}
}
$$

Error:

$$
O(h^2 + k^2)
$$

---

# 3. Aproximación de $\phi_{xx} + x\phi_{yy}$

---

## 3.1 Segunda derivada en $x$

$$
\boxed{
\left(\frac{\partial^2 \phi}{\partial x^2}\right)_{i,j}
\approx
\frac{\phi_{i+1,j} - 2\phi_{i,j} + \phi_{i-1,j}}{h^2}
}
$$

Error: $O(h^2)$

---

## 3.2 Segunda derivada en $y$

$$
\boxed{
\left(\frac{\partial^2 \phi}{\partial y^2}\right)_{i,j}
\approx
\frac{\phi_{i,j+1} - 2\phi_{i,j} + \phi_{i,j-1}}{k^2}
}
$$

Error: $O(k^2)$

---

## 3.3 Sustitución

$$
\boxed{
\left(
\frac{\partial^2 \phi}{\partial x^2}
+
x\frac{\partial^2 \phi}{\partial y^2}
\right)_{i,j}
\approx
\frac{\phi_{i+1,j} - 2\phi_{i,j} + \phi_{i-1,j}}{h^2}
+
x_i
\frac{\phi_{i,j+1} - 2\phi_{i,j} + \phi_{i,j-1}}{k^2}
}
$$

---

# 4. Resultado final

## (i)

$$
\boxed{
\frac{\partial^2 \phi}{\partial x \partial y}
\approx
\frac{
\phi_{i+1,j+1}
-\phi_{i+1,j-1}
-\phi_{i-1,j+1}
+\phi_{i-1,j-1}
}{4hk}
}
$$

---

## (ii)

$$
\boxed{
\frac{\partial^2 \phi}{\partial x^2}
+
x\frac{\partial^2 \phi}{\partial y^2}
\approx
\frac{\phi_{i+1,j} - 2\phi_{i,j} + \phi_{i-1,j}}{h^2}
+
x_i
\frac{\phi_{i,j+1} - 2\phi_{i,j} + \phi_{i,j-1}}{k^2}
}
$$

---

# 5. Conclusión

Se obtuvieron aproximaciones centradas de segundo orden usando series de Taylor.

El error global es:

$$
O(\max(h^2,k^2))
$$

como pide el problema.

# Ejercicio 2.5

Usando expansiones de Taylor alrededor de $x_0$, obtener expresiones en diferencias finitas para:

$$
\left(\frac{\partial \phi}{\partial x}\right)_{x_0}
\quad \text{y} \quad
\left(\frac{\partial^2 \phi}{\partial x^2}\right)_{x_0}
$$

correctas hasta $O(\max(h_1^2,h_2^2))$.

Además, demostrar que para $h_2 = -h_1$ se obtienen las fórmulas usuales de una malla uniforme.

---

# 1. Configuración del problema

Se consideran tres puntos en el eje $x$:

$$
x_0, \quad x_1 = x_0 + h_1, \quad x_2 = x_0 + h_2
$$

donde $h_1$ y $h_2$ no necesariamente son iguales (malla no uniforme).

Se define:

$$
\phi_0 = \phi(x_0), \quad \phi_1 = \phi(x_1), \quad \phi_2 = \phi(x_2)
$$

El objetivo es aproximar las derivadas en el punto $x_0$ utilizando los valores $\phi_0, \phi_1, \phi_2$.

---

# 2. Expansiones de Taylor

## 2.1 Expansión de $\phi_1$

Desarrollando $\phi(x_0 + h_1)$ alrededor de $x_0$:

$$
\phi_1 =
\phi_0
+ h_1 \phi'_0
+ \frac{h_1^2}{2} \phi''_0
+ \frac{h_1^3}{6} \phi'''_0
+ O(h_1^4)
$$

---

## 2.2 Expansión de $\phi_2$

De manera análoga:

$$
\phi_2 =
\phi_0
+ h_2 \phi'_0
+ \frac{h_2^2}{2} \phi''_0
+ \frac{h_2^3}{6} \phi'''_0
+ O(h_2^4)
$$

---

# 3. Construcción del sistema

Restando $\phi_0$ en ambas ecuaciones:

$$
\phi_1 - \phi_0 =
h_1 \phi'_0
+ \frac{h_1^2}{2} \phi''_0
+ \frac{h_1^3}{6} \phi'''_0
+ O(h_1^4)
$$

$$
\phi_2 - \phi_0 =
h_2 \phi'_0
+ \frac{h_2^2}{2} \phi''_0
+ \frac{h_2^3}{6} \phi'''_0
+ O(h_2^4)
$$

Estas dos ecuaciones forman un sistema que contiene las incógnitas $\phi'_0$ y $\phi''_0$.

---

# 4. Aproximación de la primera derivada

El objetivo es eliminar el término $\phi''_0$.

## Paso 1: Multiplicación

Multiplicamos:

- la primera ecuación por $h_2^2$
- la segunda ecuación por $h_1^2$

$$
h_2^2(\phi_1 - \phi_0)
=
h_1 h_2^2 \phi'_0
+ \frac{h_1^2 h_2^2}{2} \phi''_0
+ \frac{h_1^3 h_2^2}{6} \phi'''_0
$$

$$
h_1^2(\phi_2 - \phi_0)
=
h_2 h_1^2 \phi'_0
+ \frac{h_2^2 h_1^2}{2} \phi''_0
+ \frac{h_2^3 h_1^2}{6} \phi'''_0
$$

---

## Paso 2: Resta

Restamos ambas ecuaciones:

$$
h_2^2(\phi_1 - \phi_0) - h_1^2(\phi_2 - \phi_0)
$$

El término con $\phi''_0$ se cancela completamente.

Se obtiene:

$$
=
(h_1 h_2^2 - h_2 h_1^2)\phi'_0
+ O(h^3)
$$

Factorizando:

$$
h_1 h_2 (h_2 - h_1)\phi'_0 + O(h^3)
$$

---

## Resultado final

$$
\boxed{
\left(\frac{\partial \phi}{\partial x}\right)_{x_0}
=
\frac{
h_2^2(\phi_1 - \phi_0)
-
h_1^2(\phi_2 - \phi_0)
}{
h_1 h_2 (h_2 - h_1)
}
+ O(\max(h_1^2,h_2^2))
}
$$

---

# 5. Aproximación de la segunda derivada

Ahora eliminamos $\phi'_0$.

## Paso 1: Multiplicación

Multiplicamos:

- la primera ecuación por $h_2$
- la segunda ecuación por $h_1$

$$
h_2(\phi_1 - \phi_0)
=
h_1 h_2 \phi'_0
+ \frac{h_1^2 h_2}{2} \phi''_0
+ \frac{h_1^3 h_2}{6} \phi'''_0
$$

$$
h_1(\phi_2 - \phi_0)
=
h_2 h_1 \phi'_0
+ \frac{h_2^2 h_1}{2} \phi''_0
+ \frac{h_2^3 h_1}{6} \phi'''_0
$$

---

## Paso 2: Resta

Restando:

$$
h_2(\phi_1 - \phi_0) - h_1(\phi_2 - \phi_0)
$$

Los términos $\phi'_0$ se cancelan.

Se obtiene:

$$
=
\frac{h_1 h_2 (h_2 - h_1)}{2} \phi''_0
+ O(h^3)
$$

---

## Resultado final

$$
\boxed{
\left(\frac{\partial^2 \phi}{\partial x^2}\right)_{x_0}
=
\frac{
2\left[
h_2(\phi_1 - \phi_0)
-
h_1(\phi_2 - \phi_0)
\right]
}{
h_1 h_2 (h_2 - h_1)
}
+ O(\max(h_1^2,h_2^2))
}
$$

---

# 6. Caso especial: malla uniforme

Supongamos:

$$
h_2 = -h_1 = h
$$

---

## Primera derivada

$$
\left(\frac{\partial \phi}{\partial x}\right)_{x_0}
=
\frac{\phi(x_0+h) - \phi(x_0-h)}{2h}
$$

---

## Segunda derivada

$$
\left(\frac{\partial^2 \phi}{\partial x^2}\right)_{x_0}
=
\frac{\phi(x_0+h) - 2\phi(x_0) + \phi(x_0-h)}{h^2}
$$

---

# 7. Conclusión

Se han obtenido fórmulas de diferencias finitas para una malla no uniforme mediante el uso de expansiones de Taylor.

Las fórmulas:

- son válidas para cualquier $h_1$ y $h_2$
- tienen precisión de orden $O(\max(h_1^2,h_2^2))$
- se reducen correctamente a las fórmulas clásicas cuando la malla es uniforme

Esto demuestra la consistencia del método.

# Ejercicio 2.7

Se desea plantear una aproximación por diferencias finitas para el problema parabólico

$$
\frac{\partial \phi}{\partial t}
=
\frac{\partial}{\partial x}
\left[
(1+x^2)\frac{\partial \phi}{\partial x}
\right],
\qquad -1 \le x \le 1,\quad t\ge 0
$$

con condiciones:

$$
\phi(x,0)=1000-|1000x|=1000(1-|x|)
$$

y

$$
\phi(-1,t)=\phi(1,t)=0, \qquad t\ge 0
$$

tomando

$$
h=0.4, \qquad k=0.04
$$

y aprovechando la simetría para integrar hasta

$$
t=0.12.
$$

---

# 1. Interpretación del problema

La ecuación dada es una ecuación parabólica con coeficiente variable:

$$
\phi_t=\frac{\partial}{\partial x}\left((1+x^2)\phi_x\right)
$$

Obsérvese que la derivada espacial aparece en **forma conservativa**, por lo que la discretización más natural consiste en aproximar primero el flujo

$$
(1+x^2)\phi_x
$$

y después su derivada respecto de $x$.

Esto es preferible a expandir directamente la ecuación como

$$
\phi_t=(1+x^2)\phi_{xx}+2x\phi_x
$$

porque la forma conservativa conserva mejor la estructura original del problema.

---

# 2. Construcción de la malla

Tomamos la malla espacial y temporal:

$$
x_i=-1+ih, \qquad i=0,1,2,3,4,5
$$

con

$$
h=0.4
$$

Por lo tanto, los nodos espaciales son:

$$
x_0=-1,\quad x_1=-0.6,\quad x_2=-0.2,\quad x_3=0.2,\quad x_4=0.6,\quad x_5=1
$$

Para el tiempo:

$$
t_n=nk,\qquad n=0,1,2,3
$$

con

$$
k=0.04
$$

Así, los tiempos que se calcularán son:

$$
t_0=0,\qquad t_1=0.04,\qquad t_2=0.08,\qquad t_3=0.12
$$

Denotaremos:

$$
\phi_i^n \approx \phi(x_i,t_n)
$$

---

# 3. Condición inicial

La condición inicial es

$$
\phi(x,0)=1000(1-|x|)
$$

Evaluamos en los nodos:

## En $x_0=-1$

$$
\phi_0^0=1000(1-|-1|)=1000(1-1)=0
$$

## En $x_1=-0.6$

$$
\phi_1^0=1000(1-0.6)=400
$$

## En $x_2=-0.2$

$$
\phi_2^0=1000(1-0.2)=800
$$

## En $x_3=0.2$

$$
\phi_3^0=1000(1-0.2)=800
$$

## En $x_4=0.6$

$$
\phi_4^0=1000(1-0.6)=400
$$

## En $x_5=1$

$$
\phi_5^0=1000(1-1)=0
$$

Por lo tanto, la condición inicial discreta es:

$$
\boxed{
[\phi_0^0,\phi_1^0,\phi_2^0,\phi_3^0,\phi_4^0,\phi_5^0]
=
[0,400,800,800,400,0]
}
$$

---

# 4. Condiciones de frontera

Como

$$
\phi(-1,t)=0,\qquad \phi(1,t)=0
$$

entonces, para todo tiempo $n$,

$$
\phi_0^n=0,\qquad \phi_5^n=0
$$

---

# 5. Discretización de la ecuación

La ecuación es

$$
\phi_t=\frac{\partial}{\partial x}\left((1+x^2)\phi_x\right)
$$

## 5.1 Aproximación temporal

Usamos diferencia hacia adelante en el tiempo:

$$
\phi_t(x_i,t_n)\approx \frac{\phi_i^{n+1}-\phi_i^n}{k}
$$

## 5.2 Aproximación espacial en forma conservativa

Definimos

$$
a(x)=1+x^2
$$

Entonces

$$
\frac{\partial}{\partial x}\left(a(x)\phi_x\right)
$$

se aproxima por diferencias centradas como

$$
\left.\frac{\partial}{\partial x}\left(a(x)\phi_x\right)\right|_{i}^{\,n}
\approx
\frac{
a_{i+\frac12}\dfrac{\phi_{i+1}^n-\phi_i^n}{h}
-
a_{i-\frac12}\dfrac{\phi_i^n-\phi_{i-1}^n}{h}
}{h}
$$

es decir,

$$
\left.\frac{\partial}{\partial x}\left(a(x)\phi_x\right)\right|_{i}^{\,n}
\approx
\frac{
a_{i+\frac12}(\phi_{i+1}^n-\phi_i^n)
-
a_{i-\frac12}(\phi_i^n-\phi_{i-1}^n)
}{h^2}
$$

donde

$$
a_{i+\frac12}=1+x_{i+\frac12}^2,
\qquad
a_{i-\frac12}=1+x_{i-\frac12}^2
$$

y

$$
x_{i+\frac12}=x_i+\frac{h}{2},\qquad x_{i-\frac12}=x_i-\frac{h}{2}
$$

---

# 6. Esquema explícito final

Sustituyendo la aproximación temporal y espacial en la ecuación:

$$
\frac{\phi_i^{n+1}-\phi_i^n}{k}
=
\frac{
a_{i+\frac12}(\phi_{i+1}^n-\phi_i^n)
-
a_{i-\frac12}(\phi_i^n-\phi_{i-1}^n)
}{h^2}
$$

Despejando $\phi_i^{n+1}$:

$$
\boxed{
\phi_i^{n+1}
=
\phi_i^n
+
\frac{k}{h^2}
\left[
a_{i+\frac12}(\phi_{i+1}^n-\phi_i^n)
-
a_{i-\frac12}(\phi_i^n-\phi_{i-1}^n)
\right]
}
$$

Este será el esquema de diferencias finitas utilizado.

---

# 7. Cálculo del parámetro $\lambda$

Definimos

$$
\lambda=\frac{k}{h^2}
$$

Con

$$
h=0.4,\qquad k=0.04
$$

se tiene

$$
h^2=0.16
$$

y por tanto

$$
\lambda=\frac{0.04}{0.16}=0.25
$$

Así, el esquema queda:

$$
\boxed{
\phi_i^{n+1}
=
\phi_i^n
+
0.25
\left[
a_{i+\frac12}(\phi_{i+1}^n-\phi_i^n)
-
a_{i-\frac12}(\phi_i^n-\phi_{i-1}^n)
\right]
}
$$

---

# 8. Aprovechamiento de la simetría

La condición inicial

$$
\phi(x,0)=1000(1-|x|)
$$

es una función par respecto de $x=0$, es decir,

$$
\phi(-x,0)=\phi(x,0)
$$

Además:

- el coeficiente $1+x^2$ también es par,
- las condiciones de frontera en $x=-1$ y $x=1$ son iguales.

Por lo tanto, la solución conserva la simetría:

$$
\phi(-x,t)=\phi(x,t)
$$

En la malla, esto significa:

$$
\phi_1^n=\phi_4^n,\qquad \phi_2^n=\phi_3^n
$$

Así, en cada paso temporal solo necesitamos calcular dos valores interiores independientes:

$$
\phi_1^n,\qquad \phi_2^n
$$

---

# 9. Cálculo de los coeficientes intermedios

Necesitamos los valores de

$$
a_{i+\frac12}=1+x_{i+\frac12}^2
$$

## Para $i=1$

$$
x_{1+\frac12}=x_1+\frac{h}{2}=-0.6+0.2=-0.4
$$

$$
a_{1+\frac12}=1+(-0.4)^2=1+0.16=1.16
$$

$$
x_{1-\frac12}=x_1-\frac{h}{2}=-0.6-0.2=-0.8
$$

$$
a_{1-\frac12}=1+(-0.8)^2=1+0.64=1.64
$$

---

## Para $i=2$

$$
x_{2+\frac12}=x_2+\frac{h}{2}=-0.2+0.2=0
$$

$$
a_{2+\frac12}=1+0^2=1
$$

$$
x_{2-\frac12}=x_2-\frac{h}{2}=-0.2-0.2=-0.4
$$

$$
a_{2-\frac12}=1+(-0.4)^2=1.16
$$

Por simetría, para $i=3$ y $i=4$ se obtienen los mismos valores reflejados.

# Ejercicio 2.9

Se desea resolver

$$
\frac{\partial \phi}{\partial t}
=
\frac{\partial^2 \phi}{\partial x^2}
+
\frac{\partial \phi}{\partial x}
$$

en la región acotada por

$$
\phi(0,t)=0, \qquad \phi(1,t)=0, \qquad t\ge 0
$$

y con condición inicial

$$
\phi(x,0)=
\begin{cases}
x, & 0\le x\le \dfrac12,\\[0.5em]
1-x, & \dfrac12 \le x \le 1,
\end{cases}
$$

Se pide formular un esquema explícito de diferencias finitas, usando diferencias centrales en la dirección \(x\).  
Después, tomando \(h=0.1\) y un valor apropiado de \(k\), obtener una solución en \(t=0.02\).

---

# 1. Interpretación del problema

La ecuación a resolver es

$$
\phi_t=\phi_{xx}+\phi_x
$$

Esta ecuación contiene dos términos espaciales:

1. un término difusivo: \(\phi_{xx}\)
2. un término convectivo: \(\phi_x\)

Como el enunciado pide un **esquema explícito** y además especifica usar **diferencias centrales en la dirección \(x\)**, construiremos:

- diferencia hacia adelante en el tiempo para \(\phi_t\),
- diferencia central para \(\phi_{xx}\),
- diferencia central para \(\phi_x\).

---

# 2. Construcción de la malla

Tomamos una malla uniforme en el intervalo \(0\le x\le 1\):

$$
x_i=ih, \qquad i=0,1,2,\dots,10
$$

con

$$
h=0.1
$$

Por tanto, los nodos espaciales son

$$
x_0=0,\; x_1=0.1,\; x_2=0.2,\; x_3=0.3,\; x_4=0.4,\; x_5=0.5,
$$

$$
x_6=0.6,\; x_7=0.7,\; x_8=0.8,\; x_9=0.9,\; x_{10}=1.
$$

Para el tiempo, definimos

$$
t_n=nk, \qquad n=0,1,2,\dots
$$

y denotamos

$$
\phi_i^n \approx \phi(x_i,t_n).
$$

---

# 3. Aproximaciones en diferencias finitas

## 3.1 Derivada temporal

Usamos diferencia hacia adelante:

$$
\phi_t(x_i,t_n)\approx \frac{\phi_i^{n+1}-\phi_i^n}{k}.
$$

## 3.2 Segunda derivada espacial

Usamos diferencia central:

$$
\phi_{xx}(x_i,t_n)\approx
\frac{\phi_{i+1}^n-2\phi_i^n+\phi_{i-1}^n}{h^2}.
$$

## 3.3 Primera derivada espacial

Usamos también diferencia central:

$$
\phi_x(x_i,t_n)\approx
\frac{\phi_{i+1}^n-\phi_{i-1}^n}{2h}.
$$

---

# 4. Sustitución en la ecuación diferencial

Sustituyendo las tres aproximaciones en

$$
\phi_t=\phi_{xx}+\phi_x,
$$

se obtiene

$$
\frac{\phi_i^{n+1}-\phi_i^n}{k}
=
\frac{\phi_{i+1}^n-2\phi_i^n+\phi_{i-1}^n}{h^2}
+
\frac{\phi_{i+1}^n-\phi_{i-1}^n}{2h}.
$$

Multiplicando por \(k\):

$$
\phi_i^{n+1}-\phi_i^n
=
\frac{k}{h^2}\left(\phi_{i+1}^n-2\phi_i^n+\phi_{i-1}^n\right)
+
\frac{k}{2h}\left(\phi_{i+1}^n-\phi_{i-1}^n\right).
$$

Finalmente, despejando \(\phi_i^{n+1}\):

$$
\boxed{
\phi_i^{n+1}
=
\phi_i^n
+
\frac{k}{h^2}\left(\phi_{i+1}^n-2\phi_i^n+\phi_{i-1}^n\right)
+
\frac{k}{2h}\left(\phi_{i+1}^n-\phi_{i-1}^n\right)
}
$$

Este es el **esquema explícito de diferencias finitas** pedido.

---

# 5. Forma abreviada del esquema

Definimos

$$
\lambda=\frac{k}{h^2}, \qquad \mu=\frac{k}{2h}.
$$

Entonces el esquema se escribe como

$$
\boxed{
\phi_i^{n+1}
=
\phi_i^n
+
\lambda(\phi_{i+1}^n-2\phi_i^n+\phi_{i-1}^n)
+
\mu(\phi_{i+1}^n-\phi_{i-1}^n)
}
$$

o equivalentemente,

$$
\boxed{
\phi_i^{n+1}
=
(\lambda-\mu)\phi_{i-1}^n
+
(1-2\lambda)\phi_i^n
+
(\lambda+\mu)\phi_{i+1}^n
}
$$

para \(i=1,2,\dots,9\).

---

# 6. Elección de un valor apropiado de \(k\)

Como el esquema es explícito, \(k\) debe elegirse suficientemente pequeño para mantener estabilidad numérica.

Tomamos

$$
k=0.004
$$

porque:

- permite llegar exactamente a \(t=0.02\) en 5 pasos temporales,
- da un valor razonable y estable para el esquema explícito.

Con \(h=0.1\), se tiene

$$
\lambda=\frac{0.004}{(0.1)^2}=\frac{0.004}{0.01}=0.4
$$

y

$$
\mu=\frac{0.004}{2(0.1)}=\frac{0.004}{0.2}=0.02.
$$

Entonces la fórmula numérica queda

$$
\phi_i^{n+1}
=
\phi_i^n
+
0.4(\phi_{i+1}^n-2\phi_i^n+\phi_{i-1}^n)
+
0.02(\phi_{i+1}^n-\phi_{i-1}^n).
$$

Agrupando coeficientes:

$$
\phi_i^{n+1}
=
(0.4-0.02)\phi_{i-1}^n
+
(1-0.8)\phi_i^n
+
(0.4+0.02)\phi_{i+1}^n
$$

es decir,

$$
\boxed{
\phi_i^{n+1}
=
0.38\,\phi_{i-1}^n
+
0.20\,\phi_i^n
+
0.42\,\phi_{i+1}^n
}
$$

para \(i=1,\dots,9\).

---

# Ejercicio 2.12

Find the local truncation error for the Crank–Nicolson method defined by

$$
\frac{\phi_{i,j+1}-\phi_{i,j}}{k}
=
\frac{1}{2}
\left[
\frac{\phi_{i+1,j+1}-2\phi_{i,j+1}+\phi_{i-1,j+1}}{h^2}
+
\frac{\phi_{i+1,j}-2\phi_{i,j}+\phi_{i-1,j}}{h^2}
\right].
$$

Se pide encontrar el **error local de truncamiento** del método de Crank–Nicolson.

---

# 1. Idea general

El método de Crank–Nicolson se usa para aproximar la ecuación del calor

$$
\phi_t=\phi_{xx}.
$$

El error local de truncamiento se obtiene sustituyendo la **solución exacta** de la ecuación diferencial dentro del esquema numérico y desarrollando todos los términos mediante series de Taylor.

El objetivo es comparar el lado izquierdo y el lado derecho del esquema y determinar qué términos no se cancelan exactamente.

---

# 2. Notación

Sea

$$
\phi_{i,j}=\phi(x_i,t_j),
\qquad
x_i=ih,
\qquad
t_j=jk.
$$

Tomaremos como punto base el nodo \((x_i,t_j)\), y haremos expansiones de Taylor alrededor de este punto.

---

# 3. Expansión del lado izquierdo

El lado izquierdo del esquema es

$$
\frac{\phi_{i,j+1}-\phi_{i,j}}{k}.
$$

Expandimos \(\phi_{i,j+1}=\phi(x_i,t_j+k)\) alrededor de \((x_i,t_j)\):

$$
\phi_{i,j+1}
=
\phi
+k\phi_t
+\frac{k^2}{2}\phi_{tt}
+\frac{k^3}{6}\phi_{ttt}
+O(k^4).
$$

Entonces,

$$
\phi_{i,j+1}-\phi_{i,j}
=
k\phi_t
+\frac{k^2}{2}\phi_{tt}
+\frac{k^3}{6}\phi_{ttt}
+O(k^4).
$$

Dividiendo entre \(k\):

$$
\frac{\phi_{i,j+1}-\phi_{i,j}}{k}
=
\phi_t
+\frac{k}{2}\phi_{tt}
+\frac{k^2}{6}\phi_{ttt}
+O(k^3).
$$

Por lo tanto, el lado izquierdo queda

$$
\boxed{
\frac{\phi_{i,j+1}-\phi_{i,j}}{k}
=
\phi_t
+\frac{k}{2}\phi_{tt}
+\frac{k^2}{6}\phi_{ttt}
+O(k^3)
}
$$

---

# 4. Expansión del primer operador espacial en el tiempo \(t_{j+1}\)

Consideremos ahora

$$
\frac{\phi_{i+1,j+1}-2\phi_{i,j+1}+\phi_{i-1,j+1}}{h^2}.
$$

Este es el cociente central para la segunda derivada espacial, pero evaluado en el tiempo \(t_{j+1}\).

Sabemos por la fórmula central estándar que, para una función suficientemente suave,

$$
\frac{\phi_{i+1,j+1}-2\phi_{i,j+1}+\phi_{i-1,j+1}}{h^2}
=
\phi_{xx}(x_i,t_{j+1})
+\frac{h^2}{12}\phi_{xxxx}(x_i,t_{j+1})
+O(h^4).
$$

Ahora expandimos en tiempo alrededor de \(t_j\):

$$
\phi_{xx}(x_i,t_{j+1})
=
\phi_{xx}
+k\phi_{xxt}
+\frac{k^2}{2}\phi_{xxtt}
+O(k^3).
$$

Asimismo,

$$
\phi_{xxxx}(x_i,t_{j+1})
=
\phi_{xxxx}
+k\phi_{xxxxt}
+O(k^2).
$$

Por tanto,

$$
\frac{\phi_{i+1,j+1}-2\phi_{i,j+1}+\phi_{i-1,j+1}}{h^2}
=
\phi_{xx}
+k\phi_{xxt}
+\frac{k^2}{2}\phi_{xxtt}
+\frac{h^2}{12}\phi_{xxxx}
+\frac{kh^2}{12}\phi_{xxxxt}
+O(k^3)+O(h^4).
$$

---

# 5. Expansión del segundo operador espacial en el tiempo \(t_j\)

Ahora tomamos

$$
\frac{\phi_{i+1,j}-2\phi_{i,j}+\phi_{i-1,j}}{h^2}.
$$

Usando nuevamente la fórmula central para la segunda derivada:

$$
\frac{\phi_{i+1,j}-2\phi_{i,j}+\phi_{i-1,j}}{h^2}
=
\phi_{xx}
+\frac{h^2}{12}\phi_{xxxx}
+O(h^4).
$$

---

# 6. Promedio de ambos operadores espaciales

El lado derecho del esquema de Crank–Nicolson es

$$
\frac12
\left[
\frac{\phi_{i+1,j+1}-2\phi_{i,j+1}+\phi_{i-1,j+1}}{h^2}
+
\frac{\phi_{i+1,j}-2\phi_{i,j}+\phi_{i-1,j}}{h^2}
\right].
$$

Sustituyendo las dos expansiones obtenidas:

$$
\begin{aligned}
\text{Lado derecho}
&=
\frac12
\Bigg[
\phi_{xx}
+k\phi_{xxt}
+\frac{k^2}{2}\phi_{xxtt}
+\frac{h^2}{12}\phi_{xxxx}
+\frac{kh^2}{12}\phi_{xxxxt}
\\
&\qquad\qquad
+
\phi_{xx}
+\frac{h^2}{12}\phi_{xxxx}
\Bigg]
+O(k^3)+O(h^4).
\end{aligned}
$$

Agrupando términos:

$$
\text{Lado derecho}
=
\phi_{xx}
+\frac{k}{2}\phi_{xxt}
+\frac{k^2}{4}\phi_{xxtt}
+\frac{h^2}{12}\phi_{xxxx}
+\frac{kh^2}{24}\phi_{xxxxt}
+O(k^3)+O(h^4).
$$

Por lo tanto,

$$
\boxed{
\text{Lado derecho}
=
\phi_{xx}
+\frac{k}{2}\phi_{xxt}
+\frac{k^2}{4}\phi_{xxtt}
+\frac{h^2}{12}\phi_{xxxx}
+\frac{kh^2}{24}\phi_{xxxxt}
+O(k^3)+O(h^4)
}
$$

---

# 7. Comparación entre ambos lados

El esquema aproximaba la ecuación

$$
\phi_t=\phi_{xx}.
$$

Entonces, el error local de truncamiento es

$$
\tau_{i,j}
=
\frac{\phi_{i,j+1}-\phi_{i,j}}{k}
-
\frac12
\left[
\frac{\phi_{i+1,j+1}-2\phi_{i,j+1}+\phi_{i-1,j+1}}{h^2}
+
\frac{\phi_{i+1,j}-2\phi_{i,j}+\phi_{i-1,j}}{h^2}
\right].
$$

Sustituyendo las expansiones:

$$
\begin{aligned}
\tau_{i,j}
&=
\left(
\phi_t
+\frac{k}{2}\phi_{tt}
+\frac{k^2}{6}\phi_{ttt}
+O(k^3)
\right)
\\
&\quad-
\left(
\phi_{xx}
+\frac{k}{2}\phi_{xxt}
+\frac{k^2}{4}\phi_{xxtt}
+\frac{h^2}{12}\phi_{xxxx}
+\frac{kh^2}{24}\phi_{xxxxt}
+O(k^3)+O(h^4)
\right).
\end{aligned}
$$

Ahora usamos la ecuación diferencial exacta

$$
\phi_t=\phi_{xx}.
$$

Derivando respecto de \(t\):

$$
\phi_{tt}=\phi_{xxt},
$$

y derivando otra vez:

$$
\phi_{ttt}=\phi_{xxtt}.
$$

Entonces, los términos de orden \(k\) se cancelan:

$$
\frac{k}{2}\phi_{tt}-\frac{k}{2}\phi_{xxt}=0.
$$

Así,

$$
\tau_{i,j}
=
\frac{k^2}{6}\phi_{ttt}
-
\frac{k^2}{4}\phi_{xxtt}
-
\frac{h^2}{12}\phi_{xxxx}
+O(k^3)+O(h^4)+O(kh^2).
$$

Como \(\phi_{ttt}=\phi_{xxtt}\), queda

$$
\tau_{i,j}
=
\left(\frac16-\frac14\right)k^2\phi_{ttt}
-\frac{h^2}{12}\phi_{xxxx}
+O(k^3)+O(h^4)+O(kh^2).
$$

Calculamos

$$
\frac16-\frac14=-\frac1{12}.
$$

Por tanto,

$$
\boxed{
\tau_{i,j}
=
-\frac{k^2}{12}\phi_{ttt}
-\frac{h^2}{12}\phi_{xxxx}
+O(k^3)+O(h^4)+O(kh^2)
}
$$

---

# 8. Resultado final del ejercicio 2.12

El error local de truncamiento del método de Crank–Nicolson es

$$
\boxed{
\tau_{i,j}
=
-\frac{k^2}{12}\phi_{ttt}
-\frac{h^2}{12}\phi_{xxxx}
+ \text{términos de orden superior}
}
$$

Por lo tanto, el método es de orden:

- segundo orden en el tiempo,
- segundo orden en el espacio.

En notación compacta:

$$
\boxed{
\tau_{i,j}=O(k^2+h^2)
}
$$

# Ejercicio 2.13

Queremos comparar la expresión

\quad
\phi_x=\frac{\partial \phi}{\partial x}(x_i,t_j),

# 10. Expansiones de Taylor

## 10.1 Expansión de \(\phi_{i+2,j}\)

\phi_{i+2,j}
=
\phi
+2h\phi_x
+\frac{(2h)^2}{2!}\phi_{xx}

$$

---

## 10.2 Expansión de \(\phi_{i+1,j}\)

$$
\phi_{i+1,j}
=
\phi
+h\phi_x
+\frac{h^2}{2}\phi_{xx}
+\frac{h^3}{6}\phi_{xxx}
+\frac{h^4}{24}\phi_{xxxx}
+\frac{h^5}{120}\phi_{xxxxx}
+O(h^6).
$$

---

## 10.3 Expansión de \(\phi_{i-1,j}\)

$$
\phi_{i-1,j}
=
\phi
-h\phi_x
+\frac{h^2}{2}\phi_{xx}
-\frac{h^3}{6}\phi_{xxx}
+\frac{h^4}{24}\phi_{xxxx}
-\frac{h^5}{120}\phi_{xxxxx}
+O(h^6).
$$

---

## 10.4 Expansión de \(\phi_{i-2,j}\)

$$
\phi_{i-2,j}
=
\phi
-2h\phi_x
+2h^2\phi_{xx}
-\frac{4}{3}h^3\phi_{xxx}
+\frac{2}{3}h^4\phi_{xxxx}
-\frac{4}{15}h^5\phi_{xxxxx}

$$

# 11. Sustitución en la combinación dada

Consideremos

$$
\phi_{i+2,j}-\phi_{i+1,j}-\phi_{i-1,j}+\phi_{i-2,j}.
$$

Sustituyendo las expansiones anteriores:

$$
\begin{aligned}
&\phi_{i+2,j}-\phi_{i+1,j}-\phi_{i-1,j}+\phi_{i-2,j}
\\[0.5em]
&=
(\phi+2h\phi_x+2h^2\phi_{xx}+\tfrac43 h^3\phi_{xxx}+\tfrac23 h^4\phi_{xxxx}+\tfrac4{15}h^5\phi_{xxxxx})
\\
&\quad-
(\phi+h\phi_x+\tfrac12 h^2\phi_{xx}+\tfrac16 h^3\phi_{xxx}+\tfrac1{24}h^4\phi_{xxxx}+\tfrac1{120}h^5\phi_{xxxxx})
\\
&\quad-
(\phi-h\phi_x+\tfrac12 h^2\phi_{xx}-\tfrac16 h^3\phi_{xxx}+\tfrac1{24}h^4\phi_{xxxx}-\tfrac1{120}h^5\phi_{xxxxx})
\\
&\quad+
(\phi-2h\phi_x+2h^2\phi_{xx}-\tfrac43 h^3\phi_{xxx}+\tfrac23 h^4\phi_{xxxx}-\tfrac4{15}h^5\phi_{xxxxx})
+O(h^6).
\end{aligned}
$$

Ahora agrupamos término a término.

---

## 11.1 Términos con \(\phi\)

$$
\phi-\phi-\phi+\phi=0
$$

---

## 11.2 Términos con \(\phi_x\)

$$
2h\phi_x-h\phi_x+h\phi_x-2h\phi_x=0
$$

---

## 11.3 Términos con \(\phi_{xx}\)

$$
2h^2\phi_{xx}-\frac12 h^2\phi_{xx}-\frac12 h^2\phi_{xx}+2h^2\phi_{xx}
=
3h^2\phi_{xx}
$$

---

## 11.4 Términos con \(\phi_{xxx}\)

$$
\frac43 h^3\phi_{xxx}-\frac16 h^3\phi_{xxx}+\frac16 h^3\phi_{xxx}-\frac43 h^3\phi_{xxx}=0
$$

---

## 11.5 Términos con \(\phi_{xxxx}\)

$$
\frac23 h^4\phi_{xxxx}-\frac1{24}h^4\phi_{xxxx}-\frac1{24}h^4\phi_{xxxx}+\frac23 h^4\phi_{xxxx}
$$

$$
=
\left(
\frac23+\frac23-\frac1{24}-\frac1{24}
\right)h^4\phi_{xxxx}
$$

$$
=
\left(
\frac{16}{24}+\frac{16}{24}-\frac1{24}-\frac1{24}
\right)h^4\phi_{xxxx}
=
\frac{30}{24}h^4\phi_{xxxx}
=
\frac54 h^4\phi_{xxxx}
$$

---

## 11.6 Términos con \(\phi_{xxxxx}\)

Los términos impares vuelven a cancelarse por simetría, por lo tanto no contribuyen al término principal del error.

Así,

$$
\phi_{i+2,j}-\phi_{i+1,j}-\phi_{i-1,j}+\phi_{i-2,j}
=
3h^2\phi_{xx}
+\frac54 h^4\phi_{xxxx}
+O(h^6).
$$

$$
\frac{\phi_{i+2,j}-\phi_{i+1,j}-\phi_{i-1,j}+\phi_{i-2,j}}{h^2}
=
3\phi_{xx}
+\frac54 h^2\phi_{xxxx}
+O(h^4).
$$

# 12. Comparación con \(\phi_{xx}\)

La aproximación dada produce

$$
3\phi_{xx}
+\frac54 h^2\phi_{xxxx}
+O(h^4),
$$

y no directamente \(\phi_{xx}\).

Por lo tanto, si queremos aproximar \(\phi_{xx}\), primero observamos que la fórmula dada es en realidad tres veces el término principal buscado.

Así, el error local de truncamiento asociado con la expresión dada, respecto de \(\phi_{xx}\), es

$$
\left[
\frac{\phi_{i+2,j}-\phi_{i+1,j}-\phi_{i-1,j}+\phi_{i-2,j}}{h^2}
\right]
-\phi_{xx}
=
2\phi_{xx}
+\frac54 h^2\phi_{xxxx}
+O(h^4).
$$

Por consiguiente, **la fórmula dada no es una aproximación consistente de \(\phi_{xx}\)**, porque el término principal no es \(\phi_{xx}\), sino \(3\phi_{xx}\).

Si dividimos la expresión entre \(3\), obtenemos una fórmula consistente:

$$
\frac{1}{3}
\cdot
\frac{\phi_{i+2,j}-\phi_{i+1,j}-\phi_{i-1,j}+\phi_{i-2,j}}{h^2}
=
\phi_{xx}
+\frac{5}{12}h^2\phi_{xxxx}
+O(h^4).
$$

Entonces, la versión consistente tiene error local de truncamiento

$$
\boxed{
\frac{5}{12}h^2\phi_{xxxx}+O(h^4)
}
$$

---

# 14. Resultado final del ejercicio 2.13

Para la expresión dada

$$
\frac{\phi_{i+2,j}-\phi_{i+1,j}-\phi_{i-1,j}+\phi_{i-2,j}}{h^2},
$$

se obtiene

$$
\boxed{
\frac{\phi_{i+2,j}-\phi_{i+1,j}-\phi_{i-1,j}+\phi_{i-2,j}}{h^2}
=
3\phi_{xx}
+\frac54 h^2\phi_{xxxx}
+O(h^4)
}
$$

Por lo tanto:

1. La fórmula dada **no es consistente** como aproximación directa de \(\phi_{xx}\).
2. Si se corrige multiplicando por \(\frac13\), entonces sí aproxima \(\phi_{xx}\) con error de orden \(O(h^2)\).

En forma corregida:

$$
\boxed{
\frac{1}{3}
\cdot
\frac{\phi_{i+2,j}-\phi_{i+1,j}-\phi_{i-1,j}+\phi_{i-2,j}}{h^2}
=
\phi_{xx}
+\frac{5}{12}h^2\phi_{xxxx}
+O(h^4)
}
$$

# 15. Conclusión general

En el ejercicio 2.12 se verificó que el método de Crank–Nicolson tiene error local de truncamiento

$$
O(k^2+h^2),
$$

lo cual confirma que es de segundo orden tanto en tiempo como en espacio.

En el ejercicio 2.13 se comprobó que la fórmula propuesta no aproxima directamente \(\phi_{xx}\), sino \(3\phi_{xx}\).  
Por ello, tal como está escrita, no es consistente. Sin embargo, una versión corregida obtenida dividiendo entre \(3\) sí proporciona una aproximación de segundo orden para la derivada segunda espacial.

\frac{\phi_{i,j+2}+2\phi_{i,j+1}-2\phi_{i,j-1}+\phi_{i,j-2}}{8k}.
$$

Como los subíndices cambian únicamente en la dirección $j$, esto significa que la aproximación actúa en la variable asociada al paso $k$, es decir, en la segunda variable de la malla.

Por lo tanto, debemos desarrollar mediante series de Taylor los términos

$$
\phi_{i,j+2}, \qquad \phi_{i,j+1}, \qquad \phi_{i,j-1}, \qquad \phi_{i,j-2}
$$

alrededor del punto $(i,j)$.

Para simplificar la notación, escribiremos

$$
\phi = \phi(x_i,t_j),
$$

y sus derivadas respecto de $t$ como

$$
\phi_t,\quad \phi_{tt},\quad \phi_{ttt},\quad \phi_{tttt},\quad \phi_{ttttt}, \dots
$$

todas evaluadas en el punto $(x_i,t_j)$.

---

# 2. Expansiones de Taylor

## 2.1 Expansión de $\phi_{i,j+2}$

Desarrollamos $\phi(x_i,t_j+2k)$ alrededor de $(x_i,t_j)$:

$$
\phi_{i,j+2}
=
\phi
+2k\phi_t
+\frac{(2k)^2}{2!}\phi_{tt}
+\frac{(2k)^3}{3!}\phi_{ttt}
+\frac{(2k)^4}{4!}\phi_{tttt}
+\frac{(2k)^5}{5!}\phi_{ttttt}
+O(k^6).
$$

Simplificando coeficientes:

$$
\phi_{i,j+2}
=
\phi
+2k\phi_t
+2k^2\phi_{tt}
+\frac{4}{3}k^3\phi_{ttt}
+\frac{2}{3}k^4\phi_{tttt}
+\frac{4}{15}k^5\phi_{ttttt}
+O(k^6).
$$

---

## 2.2 Expansión de $\phi_{i,j+1}$

Ahora desarrollamos $\phi(x_i,t_j+k)$:

$$
\phi_{i,j+1}
=
\phi
+k\phi_t
+\frac{k^2}{2}\phi_{tt}
+\frac{k^3}{6}\phi_{ttt}
+\frac{k^4}{24}\phi_{tttt}
+\frac{k^5}{120}\phi_{ttttt}
+O(k^6).
$$

---

## 2.3 Expansión de $\phi_{i,j-1}$

Desarrollando $\phi(x_i,t_j-k)$:

$$
\phi_{i,j-1}
=
\phi
-k\phi_t
+\frac{k^2}{2}\phi_{tt}
-\frac{k^3}{6}\phi_{ttt}
+\frac{k^4}{24}\phi_{tttt}
-\frac{k^5}{120}\phi_{ttttt}
+O(k^6).
$$

---

## 2.4 Expansión de $\phi_{i,j-2}$

Finalmente, desarrollamos $\phi(x_i,t_j-2k)$:

$$
\phi_{i,j-2}
=
\phi
-2k\phi_t
+2k^2\phi_{tt}
-\frac{4}{3}k^3\phi_{ttt}
+\frac{2}{3}k^4\phi_{tttt}
-\frac{4}{15}k^5\phi_{ttttt}
+O(k^6).
$$

---

# 3. Sustitución en la expresión dada

La expresión del problema es

$$
\frac{\phi_{i,j+2}+2\phi_{i,j+1}-2\phi_{i,j-1}+\phi_{i,j-2}}{8k}.
$$

Primero analizamos el numerador:

$$
\phi_{i,j+2}+2\phi_{i,j+1}-2\phi_{i,j-1}+\phi_{i,j-2}.
$$

Sustituyendo las expansiones de Taylor:

$$
\begin{aligned}
&\phi_{i,j+2}+2\phi_{i,j+1}-2\phi_{i,j-1}+\phi_{i,j-2}
\\[0.5em]
&=
\left(
\phi
+2k\phi_t
+2k^2\phi_{tt}
+\frac{4}{3}k^3\phi_{ttt}
+\frac{2}{3}k^4\phi_{tttt}
+\frac{4}{15}k^5\phi_{ttttt}
\right)
\\
&\quad+
2\left(
\phi
+k\phi_t
+\frac{k^2}{2}\phi_{tt}
+\frac{k^3}{6}\phi_{ttt}
+\frac{k^4}{24}\phi_{tttt}
+\frac{k^5}{120}\phi_{ttttt}
\right)
\\
&\quad-
2\left(
\phi
-k\phi_t
+\frac{k^2}{2}\phi_{tt}
-\frac{k^3}{6}\phi_{ttt}
+\frac{k^4}{24}\phi_{tttt}
-\frac{k^5}{120}\phi_{ttttt}
\right)
\\
&\quad+
\left(
\phi
-2k\phi_t
+2k^2\phi_{tt}
-\frac{4}{3}k^3\phi_{ttt}
+\frac{2}{3}k^4\phi_{tttt}
-\frac{4}{15}k^5\phi_{ttttt}
\right)
+O(k^6).
\end{aligned}
$$

Ahora agrupamos término por término.

---

# 4. Agrupación de términos

## 4.1 Términos con $\phi$

Los coeficientes de $\phi$ son

$$
1+2-2+1=2.
$$

Por lo tanto, el término constante es

$$
2\phi.
$$

---

## 4.2 Términos con $\phi_t$

Los coeficientes de $k\phi_t$ son:

- de $\phi_{i,j+2}$: $2$
- de $2\phi_{i,j+1}$: $2$
- de $-2\phi_{i,j-1}$: $+2$
- de $\phi_{i,j-2}$: $-2$

Entonces,

$$
2+2+2-2=4.
$$

Por tanto, el término lineal es

$$
4k\phi_t.
$$

---

## 4.3 Términos con $\phi_{tt}$

Los coeficientes de $k^2\phi_{tt}$ son:

- de $\phi_{i,j+2}$: $2$
- de $2\phi_{i,j+1}$: $2\cdot \frac12 = 1$
- de $-2\phi_{i,j-1}$: $-2\cdot \frac12 = -1$
- de $\phi_{i,j-2}$: $2$

Sumando:

$$
2+1-1+2=4.
$$

Así aparece el término

$$
4k^2\phi_{tt}.
$$

---

## 4.4 Términos con $\phi_{ttt}$

Los coeficientes de $k^3\phi_{ttt}$ son:

- de $\phi_{i,j+2}$: $\frac43$
- de $2\phi_{i,j+1}$: $2\cdot \frac16 = \frac13$
- de $-2\phi_{i,j-1}$: $-2\cdot \left(-\frac16\right)=\frac13$
- de $\phi_{i,j-2}$: $-\frac43$

Entonces,

$$
\frac43+\frac13+\frac13-\frac43=\frac23.
$$

Por tanto, aparece el término

$$
\frac23 k^3\phi_{ttt}.
$$

---

## 4.5 Términos con $\phi_{tttt}$

Los coeficientes de $k^4\phi_{tttt}$ son:

- de $\phi_{i,j+2}$: $\frac23$
- de $2\phi_{i,j+1}$: $2\cdot \frac1{24}=\frac1{12}$
- de $-2\phi_{i,j-1}$: $-2\cdot \frac1{24}=-\frac1{12}$
- de $\phi_{i,j-2}$: $\frac23$

Entonces,

$$
\frac23+\frac1{12}-\frac1{12}+\frac23=\frac43.
$$

Así obtenemos el término

$$
\frac43 k^4\phi_{tttt}.
$$

---

## 4.6 Términos con $\phi_{ttttt}$

Los coeficientes de $k^5\phi_{ttttt}$ son:

- de $\phi_{i,j+2}$: $\frac4{15}$
- de $2\phi_{i,j+1}$: $2\cdot \frac1{120}=\frac1{60}$
- de $-2\phi_{i,j-1}$: $-2\cdot \left(-\frac1{120}\right)=\frac1{60}$
- de $\phi_{i,j-2}$: $-\frac4{15}$

Entonces,

$$
\frac4{15}+\frac1{60}+\frac1{60}-\frac4{15}
=
\frac1{30}.
$$

Por lo tanto, aparece el término

$$
\frac1{30}k^5\phi_{ttttt}.
$$

---

# 5. Expresión completa del numerador

Finalmente,

$$
\phi_{i,j+2}+2\phi_{i,j+1}-2\phi_{i,j-1}+\phi_{i,j-2}
=
2\phi
+4k\phi_t
+4k^2\phi_{tt}
+\frac23 k^3\phi_{ttt}
+\frac43 k^4\phi_{tttt}
+\frac1{30}k^5\phi_{ttttt}
+O(k^6).
$$

Dividiendo entre $8k$:

$$
\frac{\phi_{i,j+2}+2\phi_{i,j+1}-2\phi_{i,j-1}+\phi_{i,j-2}}{8k}
=
\frac{2\phi}{8k}
+\frac{4k\phi_t}{8k}
+\frac{4k^2\phi_{tt}}{8k}
+\frac{\frac23 k^3\phi_{ttt}}{8k}
+\frac{\frac43 k^4\phi_{tttt}}{8k}
+\cdots
$$

es decir,

$$
=
\frac{\phi}{4k}
+\frac12\phi_t
+\frac12 k\phi_{tt}
+\frac1{12}k^2\phi_{ttt}
+\frac16 k^3\phi_{tttt}
+\frac1{240}k^4\phi_{ttttt}
+O(k^5).
$$

---

# 6. Interpretación del resultado

El desarrollo obtenido contiene el término

$$
\frac{\phi}{4k},
$$

el cual no debería aparecer si la fórmula fuera una aproximación consistente de una derivada parcial.

En una aproximación correcta de una derivada, el término dominante debe ser precisamente la derivada buscada. Aquí, en cambio, aparece un término proporcional al valor de la función misma dividido entre $k$.

Esto significa que la fórmula dada, tal como aparece en el enunciado,

$$
\frac{\phi_{i,j+2}+2\phi_{i,j+1}-2\phi_{i,j-1}+\phi_{i,j-2}}{8k}
$$

**no es una aproximación consistente de una derivada parcial**.

---

# 7. Comparación formal con $\phi_t$

Si, de manera formal, comparamos la expresión con $\phi_t$, obtenemos

$$
\begin{aligned}
\frac{\phi_{i,j+2}+2\phi_{i,j+1}-2\phi_{i,j-1}+\phi_{i,j-2}}{8k}
-\phi_t
&=
\frac{\phi}{4k}
-\frac12\phi_t
+\frac12 k\phi_{tt}
+\frac1{12}k^2\phi_{ttt}
\\
&\quad
+\frac16 k^3\phi_{tttt}
+O(k^4).
\end{aligned}
$$

El término dominante del error sería entonces

$$
\frac{\phi}{4k},
$$

lo cual confirma que la fórmula no es consistente.

---

# 8. Observación sobre posible error tipográfico

En este tipo de ejercicios es común encontrar fórmulas centradas de orden superior para la primera derivada, por ejemplo

$$
\frac{-\phi_{i,j+2}+8\phi_{i,j+1}-8\phi_{i,j-1}+\phi_{i,j-2}}{12k},
$$

que sí aproxima consistentemente a $\phi_t$.

Por eso, es muy probable que la expresión del enunciado contenga un error tipográfico. Sin embargo, **tomando literalmente la fórmula dada**, la conclusión correcta es que no aproxima ninguna derivada parcial de manera consistente.

---

# 9. Resultado final

La expresión

$$
\frac{\phi_{i,j+2}+2\phi_{i,j+1}-2\phi_{i,j-1}+\phi_{i,j-2}}{8k}
$$

tiene el desarrollo de Taylor

$$
\boxed{
\frac{\phi_{i,j+2}+2\phi_{i,j+1}-2\phi_{i,j-1}+\phi_{i,j-2}}{8k}
=
\frac{\phi}{4k}
+\frac12\phi_t
+\frac12 k\phi_{tt}
+\frac1{12}k^2\phi_{ttt}
+\frac16 k^3\phi_{tttt}
+O(k^4)
}
$$

Por consiguiente:

1. No aproxima consistentemente una derivada parcial.
2. El término principal no es una derivada, sino $\dfrac{\phi}{4k}$.
3. La fórmula es inconsistente tal como está escrita.

---

# 10. Conclusión

Mediante expansiones de Taylor detalladas alrededor del punto $(i,j)$ se demostró que la expresión dada no constituye una aproximación consistente de una derivada parcial, ya que aparece un término dominante proporcional a $\phi/(4k)$.

Por ello, la respuesta correcta, tomando el enunciado literalmente, es que la fórmula no aproxima ninguna derivada parcial de forma válida y que su “error de truncamiento” confirma la inconsistencia del esquema.

# Ejercicio 2.19

Considérese nuevamente el Ejercicio 2.9, en el cual la ecuación diferencial parcial

$$
\frac{\partial \phi}{\partial t}
=
\frac{\partial^2 \phi}{\partial x^2}
+
\frac{\partial \phi}{\partial x}
$$

se resuelve mediante un método explícito.  
Se pide usar los teoremas de Gershgorin para examinar la estabilidad del método.

---

# 1. Objetivo del ejercicio

En el Ejercicio 2.9 se construyó un esquema explícito para la ecuación

$$
\phi_t=\phi_{xx}+\phi_x
$$

usando:

- diferencia hacia adelante en el tiempo,
- diferencias centrales en la variable espacial $x$.

Ahora no se pide volver a construir el esquema, sino **analizar su estabilidad** usando los **teoremas de Gershgorin** aplicados a la matriz de iteración del método explícito.

La idea central es la siguiente:

- escribir el esquema en forma matricial,
- identificar la matriz de avance temporal,
- localizar sus autovalores usando los discos de Gershgorin,
- imponer condiciones para que dichos autovalores queden dentro de la región de estabilidad.

---

# 2. Recordatorio del esquema explícito del Ejercicio 2.9

La ecuación es

$$
\phi_t=\phi_{xx}+\phi_x.
$$

Usamos las aproximaciones estándar:

## Derivada temporal

$$
\phi_t(x_i,t_n)\approx \frac{\phi_i^{n+1}-\phi_i^n}{k}
$$

## Segunda derivada espacial

$$
\phi_{xx}(x_i,t_n)\approx
\frac{\phi_{i+1}^n-2\phi_i^n+\phi_{i-1}^n}{h^2}
$$

## Primera derivada espacial

$$
\phi_x(x_i,t_n)\approx
\frac{\phi_{i+1}^n-\phi_{i-1}^n}{2h}
$$

Sustituyendo en la ecuación diferencial:

$$
\frac{\phi_i^{n+1}-\phi_i^n}{k}
=
\frac{\phi_{i+1}^n-2\phi_i^n+\phi_{i-1}^n}{h^2}
+
\frac{\phi_{i+1}^n-\phi_{i-1}^n}{2h}.
$$

Multiplicando por $k$ y despejando $\phi_i^{n+1}$:

$$
\phi_i^{n+1}
=
\phi_i^n
+
\frac{k}{h^2}\left(\phi_{i+1}^n-2\phi_i^n+\phi_{i-1}^n\right)
+
\frac{k}{2h}\left(\phi_{i+1}^n-\phi_{i-1}^n\right).
$$

Definimos

$$
\lambda=\frac{k}{h^2}, \qquad \mu=\frac{k}{2h}.
$$

Entonces el esquema puede escribirse como

$$
\boxed{
\phi_i^{n+1}
=
(\lambda-\mu)\phi_{i-1}^n
+
(1-2\lambda)\phi_i^n
+
(\lambda+\mu)\phi_{i+1}^n
}
$$

para los nodos interiores.

---

# 3. Forma matricial del método

Sea

$$
U^n=
\begin{pmatrix}
\phi_1^n\\
\phi_2^n\\
\vdots\\
\phi_{m-1}^n
\end{pmatrix},
$$

el vector de valores interiores en el tiempo $t_n$.

Entonces el método explícito puede escribirse en forma matricial como

$$
U^{n+1}=A\,U^n,
$$

donde la matriz $A$ es tridiagonal:

$$
A=
\begin{pmatrix}
1-2\lambda & \lambda+\mu & 0 & \cdots & 0\\
\lambda-\mu & 1-2\lambda & \lambda+\mu & \ddots & \vdots\\
0 & \lambda-\mu & 1-2\lambda & \ddots & 0\\
\vdots & \ddots & \ddots & \ddots & \lambda+\mu\\
0 & \cdots & 0 & \lambda-\mu & 1-2\lambda
\end{pmatrix}.
$$

La estabilidad del método explícito depende de que las potencias de esta matriz no crezcan sin control cuando avanzamos en el tiempo.

Una condición suficiente natural es exigir que todos los autovalores de $A$ satisfagan

$$
|\rho|\le 1.
$$

Para examinar esto, usaremos el teorema de Gershgorin.

---

# 4. Teorema de Gershgorin

El teorema de Gershgorin afirma que todo autovalor $\rho$ de una matriz cuadrada pertenece al menos a uno de los discos de Gershgorin:

$$
|\rho-a_{ii}|
\le
\sum_{j\ne i}|a_{ij}|.
$$

En nuestro caso, cada fila interior de la matriz $A$ tiene:

- elemento diagonal

$$
a_{ii}=1-2\lambda,
$$

- elementos fuera de la diagonal

$$
a_{i,i-1}=\lambda-\mu,
\qquad
a_{i,i+1}=\lambda+\mu.
$$

Por tanto, el radio del disco de Gershgorin correspondiente es

$$
R=|\lambda-\mu|+|\lambda+\mu|.
$$

Entonces, todos los autovalores de $A$ están contenidos en los discos centrados en

$$
1-2\lambda
$$

con radio

$$
|\lambda-\mu|+|\lambda+\mu|.
$$

Es decir,

$$
\boxed{
|\rho-(1-2\lambda)|
\le
|\lambda-\mu|+|\lambda+\mu|
}
$$

---

# 5. Condición para la estabilidad

Para que el método explícito sea estable, queremos que todos los discos de Gershgorin queden contenidos dentro del intervalo real $[-1,1]$, o más generalmente dentro del disco unidad en el plano complejo.

Como en este caso la matriz es real tridiagonal y los discos están centrados sobre el eje real, una condición suficiente es que el extremo izquierdo no sea menor que $-1$ y el extremo derecho no sea mayor que $1$.

Esto significa:

## Extremo derecho

$$
(1-2\lambda)+\left(|\lambda-\mu|+|\lambda+\mu|\right)\le 1
$$

## Extremo izquierdo

$$
(1-2\lambda)-\left(|\lambda-\mu|+|\lambda+\mu|\right)\ge -1
$$

---

# 6. Simplificación bajo la hipótesis natural $\lambda \ge |\mu|$

En la práctica numérica, interesa el caso en el cual

$$
\lambda \ge |\mu|.
$$

Eso equivale a

$$
\frac{k}{h^2}\ge \frac{k}{2h}.
$$

Si $k>0$, esto se simplifica como

$$
\frac1{h^2}\ge \frac1{2h}
\quad\Longleftrightarrow\quad
2\ge h.
$$

Como usualmente $h<1$, esta condición se cumple sin dificultad.

Bajo esta hipótesis:

$$
\lambda+\mu \ge 0,
\qquad
\lambda-\mu \ge 0,
$$

y por tanto

$$
|\lambda-\mu|+|\lambda+\mu|
=
(\lambda-\mu)+(\lambda+\mu)=2\lambda.
$$

Así, el disco de Gershgorin se reduce a

$$
|\rho-(1-2\lambda)|\le 2\lambda.
$$

Es decir, los autovalores están en el intervalo

$$
[\,1-4\lambda,\;1\,].
$$

---

# 7. Condición resultante

Para que este intervalo quede dentro de $[-1,1]$, basta exigir:

## Límite superior

El extremo derecho es $1$, así que automáticamente cumple.

## Límite inferior

Necesitamos

$$
1-4\lambda \ge -1.
$$

Entonces,

$$
-4\lambda \ge -2
$$

y por tanto

$$
\lambda \le \frac12.
$$

Recordando que

$$
\lambda=\frac{k}{h^2},
$$

obtenemos la condición de estabilidad

$$
\boxed{
\frac{k}{h^2}\le \frac12
}
$$

o equivalentemente

$$
\boxed{
k\le \frac{h^2}{2}.
}
$$

Esta es precisamente la condición clásica de estabilidad del esquema explícito para la ecuación del calor, y aquí sigue siendo válida en este caso con el término convectivo tratado con diferencia central.

---

# 8. Interpretación del papel del término convectivo

El término adicional $\phi_x$ introduce la contribución $\mu$ en la matriz de iteración. Sin embargo, bajo la hipótesis $\lambda \ge |\mu|$, los radios de Gershgorin siguen siendo iguales a $2\lambda$.

Esto muestra que el término convectivo no altera la condición principal de estabilidad obtenida por Gershgorin, siempre que la malla satisfaga la relación usual y que el paso espacial no sea demasiado grande.

En consecuencia, la restricción dominante sigue siendo

$$
\frac{k}{h^2}\le \frac12.
$$

---

# 9. Aplicación al caso del Ejercicio 2.9

En el Ejercicio 2.9 se tomó

$$
h=0.1.
$$

Entonces

$$
h^2=0.01.
$$

La condición de estabilidad exige

$$
k\le \frac{0.01}{2}=0.005.
$$

Por lo tanto, cualquier elección de $k$ que satisfaga

$$
k\le 0.005
$$

es estable según este análisis de Gershgorin.

En el ejercicio anterior se eligió

$$
k=0.004,
$$

y claramente

$$
0.004<0.005.
$$

Así, esa elección satisface la condición de estabilidad.

---

# 10. Verificación numérica con $h=0.1$ y $k=0.004$

Calculamos

$$
\lambda=\frac{k}{h^2}=\frac{0.004}{0.01}=0.4,
$$

y

$$
\mu=\frac{k}{2h}=\frac{0.004}{0.2}=0.02.
$$

Entonces:

$$
\lambda-\mu=0.38,
\qquad
\lambda+\mu=0.42.
$$

La diagonal de la matriz es

$$
1-2\lambda=1-0.8=0.2.
$$

El radio del disco de Gershgorin es

$$
R=|\lambda-\mu|+|\lambda+\mu|=0.38+0.42=0.8.
$$

Por tanto, los autovalores están contenidos en

$$
|\rho-0.2|\le 0.8,
$$

es decir, dentro del intervalo

$$
[-0.6,\,1].
$$

Este intervalo está completamente contenido en $[-1,1]$, lo cual confirma la estabilidad del método para estos valores.

---

# 11. Resultado final

Usando los teoremas de Gershgorin, se concluye que todos los autovalores de la matriz de iteración del esquema explícito satisfacen

$$
|\rho-(1-2\lambda)|
\le
|\lambda-\mu|+|\lambda+\mu|.
$$

Bajo la condición natural $\lambda \ge |\mu|$, esto se reduce a

$$
|\rho-(1-2\lambda)|\le 2\lambda,
$$

por lo que los autovalores quedan en el intervalo

$$
[1-4\lambda,\,1].
$$

Para que el método sea estable, basta exigir

$$
1-4\lambda \ge -1,
$$

lo que conduce a

$$
\boxed{
\lambda \le \frac12
}
$$

o equivalentemente

$$
\boxed{
\frac{k}{h^2}\le \frac12.
}
$$

En el caso particular del Ejercicio 2.9, con

$$
h=0.1,\qquad k=0.004,
$$

se tiene

$$
\lambda=0.4<0.5,
$$

por lo que el método es estable.

---

# 12. Conclusión

El análisis mediante los discos de Gershgorin muestra que el esquema explícito asociado a

$$
\phi_t=\phi_{xx}+\phi_x
$$

es estable si se cumple la condición

$$
\frac{k}{h^2}\le \frac12.
$$

Aplicando esta condición al caso numérico del Ejercicio 2.9, se verifica que la elección

$$
h=0.1,\qquad k=0.004
$$

es adecuada, ya que garantiza que todos los autovalores de la matriz de iteración permanezcan dentro de la región estable.
